요즘은 전세계적으로 TV 안보는 사람은 있어도 유튜브 안보는 사람은 없을 정도로 많은 사람들이 유튜브에서 동영상 시청을 하는 세상입니다. 이번 강좌에서는 유튜브에서 동영상을 어떻게 다운로드 하는지에 대해서 알아보도록 하겠습니다. 일단 유튜브의 URL 을 살펴보면 위의 이미지에서 처럼 ```watch``` 라는 주소에 ```v``` 라는 매개변수명에 ```3-mHctyi3_Y``` 이런 값을 넘겨주는것으로 보아 ```3-mHctyi3_Y``` 값이 동영상의 고유값이라는 사실을 유추할 수 있습니다.

<font color="red" style="background-color:#eeeeee;">웹사이트나 페이지의 코드는 언제든 변경될 수 있습니다. 따라서 크롤링의 특성상 어떻게 사이트를 분석하고 다운로드 하는지 기본적인 이해도가 있어야만 그런 변화된 환경에 대처할 수 있습니다.</font>

보통 웹에서 특별한 프로그램없이 브라우저 자체 기능만으로 동영상을 재생할때는 일반적으로 위의 코드처럼 ```video``` 태그를 작성하고 ```src=동영상주소``` 를 작성하면 HTML5 가 지원하는 브라우저에서는 자동으로 동영상이 재생되는게 일반적입니다.

![img](images/14.jpg)

크롬으로 유튜브 동영상 재생 페이지를 열어 개발자도구 (F12키) 로 확인을 해보면 유튜브 역시 ```<video>``` 태그가 존재하고 ```src``` 에 영상의 주소가 있는것을 확인할 수 있습니다. 해당 ```src``` 의 주소에서 간단하게 동영상을 다운로드 하면 되는 간단한 문제입니다만 막상 코딩을 하다보면 전혀 간단한 문제가 아니란 사실을 금방 알수 있습니다.

![img](images/15.jpg)

크롬 개발자모드에서 Elements 탭이 아닌 Network 탭에서 다시 새로고침을 해보면 위의 이미지에서 처럼 우리는 ```watch``` 페이지 하나를 새로고침했을 뿐인데 자바스크립트 파일, 스타일시트, 로깅, 이미지 등등등 실제 122번의 요청이 일어난것을 알 수 있습니다. 그 중 맨 위의 동영상 재생 주소인 ```watch``` 주소를 살펴보면 실제 크롬 브라우저가 유튜브 서버에서 ```watch``` 주소를 접속하면서 받은 Response 데이터를 살펴볼수 있습니다. 이 Network 탭에서 CTRL + F 키를 눌러 ```<video``` 로 시작하는 태그를 검색해보면 검색 결과가 0 으로 나오는것을 확인할 수 있습니다.

개발자도구로 살펴본 Element 탭은 현재 우리가 눈으로 보고 있는 최종 렌더링 결과물 입니다. 그러나 서버로부터 받은 데이터는 브라우저에 의해 렌더링 되기전의 데이터기 때문에 이와 다를수 있습니다. 그 말은 서버로부터 받은 데이터와 우리가 눈으로 보는 최종 결과물 사이에 어떤 작업이 이뤄졌다는 이야기가 되고 이 작업은 거의 대부분 자바스크립에 의해서 이루어집니다. 그렇기 때문에 우리는 실제 동영상을 볼 수 있게 Element 탭에서는 ```<video>``` 태그가 보였지만 서버로부터 다운로드 한 Network 탭의 Response 데이터에서는 ```<video>``` 태그가 보이지 않는것 입니다. 그럼 우리는 Response 한 데이터부터 분석을 해서 ```<video>``` 태그와 실제 동영상 주소인 ```src``` 의 값을 찾아내야 합니다.

일단 Network 탭의 Response 데이터를 개발자도구를 통해 보기에 불편하기에 Reponse 창에서 마우스 우클릭하여 Save as... 를 해서 텍스트 파일로 저장 후에 컴퓨터의 텍스트 뷰어 / 에디터 프로그램을 이용해서 분석을 하는게 좋습니다. 여기서는 [[서브라임텍스트]](https://www.sublimetext.com/3) 를 사용하도록 하겠습니다.

위 이미지는 서브라임텍스트로 ```watch``` 페이지의 Response 데이터를 텍스트파일로 저장하여 읽은 상태입니다. 보통 이런 사이트들은 여러가지 이유로 최대한 난독화가 되어있기 때문에 인내와 끈기가 필요한 부분이기도 합니다. 위의 이미지에서 라인넘버를 보면 모든 코드가 한줄에 들어가있는 경우도 있습니다. 사이트를 분석할때 경우에 따라서는 현재 페이지에 링크되어있는 자바스크립트 모두를 분석해야 하는 경우도 있습니다.


![img](images/22.jpg)

우리는 방대하고 난독화 되어있는 Response 데이터를 무작정 분석할 수 없기 때문에 최소한의 힌트를 찾아내야 합니다. 일단 크롬의 개발자 도구의 Network 탭을 열어놓은 상태에서 페이지를 새로고침하고 영상을 재생시켜놓으면 계속해서 트래픽이 발생하는것을 확인할 수 있고 **videoplayback** 으로 시작하는 주소가 xhr 통신으로 계속해서 반복적으로 로딩되는것을 확인할 수 있습니다. xhr 은 XMLHttpRequest 로 서버와 브라우저가 내부적인 통신을 하고 있는 형태를 의미합니다.

![img](images/23.jpg)

**videoplayback** 항목을 클릭해서 우측의 Headers 탭을 살펴보면 해당 videoplayback 의 Request URL 을 확인할 수 있습니다. 이 Request URL 이 실제 접속되고 있는 서버의 주소 입니다. 아마도 이 주소가 동영상 주소로 추측됩니다. 일단 이 문자열을 적당히 드래그 하여 CTRL + C 를 눌러 클립보드로 복사해놓습니다.


![img](images/24.jpg)

아까 위에서 저장해놓은 ```watch.txt``` 파일에서 방금 복사한 주소를 찾을 수 있습니다. 이 주소 값은 다른 여러가지 값들과 함께 자바스크립트의 ```ytInitialPlayerResponse``` 라는 이름의 변수에 저장되어 있는 값으로 확인 됩니다. 자바스크립트에서는 변수 선언이 ```var``` 로 시작됩니다. 이 변수 값에는 {  로 시작하는것으로 보아 json 형태의 값으로 뭔가 알 수 없지만 많은 키와 값의 데이터가 선언된것을 확인 할 수 있습니다. 이 값에는 무엇인지는 모르지만 숫자, 키, 값 등등이 있고 무언가 우리가 조금은 익숙한 format, playbackMode 이런 단어들도 보이는것 같은데 이 자체로만은 알아보기가 너무 복잡하니 파이썬으로 이 값만 추려서 확인을 해봐야 할듯 합니다.

````python
from urllib.request import Request, urlopen
import json
import pprint

header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36",
    "Referer": "https://youtube.com"
}

video_id = "3-mHctyi3_Y"
watch_url = "https://www.youtube.com/watch?v={}".format(video_id)

req = Request(watch_url, headers=header)
watch_html = urlopen(req).read().decode("utf-8")
start_str = "ytInitialPlayerResponse = "
start_index = watch_html.find(start_str)
end_index = watch_html.find("};", start_index + 1) if start_index >= 0 else 0

if start_index < end_index:
    h = watch_html[start_index + len(start_str):end_index + 1]
    _json = json.loads(h)
    pprint.pprint(_json)
    with open('output.txt', 'wt') as out:
        pprint.pprint(_json, stream=out)
````

위의 코드는 ```watch``` 페이지를 다운로드 하여 문자열 데이터에서 ```ytInitialPlayerResponse = ``` 의 내용만 추출하는 코드 입니다. 이렇게 ```requests.get()``` 처럼 서버에 페이지를 요청하게 되면 우리는 크롬 개발자도구 Network 탭의 Response 항목에서처럼 서버가 전송해주는 데이터를 받을 수 있습니다. <font color="red" style="background-color:#eeeeee;">이 값은 브라우저에서 최종적으로 보여지는 렌더링 된 내용과는 다르다는 사실을 항상 염두해야 합니다.</font>

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">

    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36",

유튜브는 모바일, 태블릿 등 다양한 기기를 수용하기 때문에 컴퓨터 환경에서 접속한것 처럼 서버에 알려주기 위해 헤더에 ```User-Agent``` 값을 설정해서 ```https://www.youtube.com/watch?v=3-mHctyi3_Y``` 주소로 접속하여 받은 데이터를 ```watch_html``` 에 문자열(text) 형태로 저장합니다. ```Referer``` 는 현재 페이지를 접속하기 바로 이전에 어떤 URL을 들렸다 왔는지를 참조하는 내용 입니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
start_str = "ytInitialPlayerResponse = "
start_index = watch_html.find(start_str)
</pre>

```ytInitialPlayerResponse = ``` 라는 문자열 값을 ```start_str``` 변수에 저장합니다. ```watch_html.find()``` 함수를 사용하여 ```start_str``` 문자열 변수의 내용을 검색합니다. 이때 검색 결과가 존재하면 해당 문자열 시작 위치 값이 ```start_index``` 변수에 저장되고 검색 결과가 없으면 -1 이 저장됩니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
end_index = watch_html.find("};", start_index + 1) if start_index >= 0 else 0
</pre>

```start_index``` 변수 값이 0보다 크거나 같으면 ```watch_html``` 에서 ```"};"``` 문자열을 검색하는데 자바스크립에서 JSON 을 변수로 선언할때는 ```var 변수명 = {};``` 이런식으로 선언하기 때문입니다. 이때 여기서 중요한 부분은 ```watch_html``` 에서 "};" 를 찾는 시작위치는 ```변수명 = {``` 를 찾은 위치부터 검색해야 뒤에 나오는 ```"};"``` 를 찾을 수 있습니다. 물론 정규식 같은 방법을 사용해서 더 편하게 원하는 내용을 찾을 수 있겠습니다만 여기서는 이해를 돕기위해 가장 단순한 파싱 기법을 사용했습니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
if start_index < end_index:
    h = watch_html[start_index + len(start_str):end_index + 1]
    _json = json.loads(h)
    pprint.pprint(_json)
</pre>

그렇게 우리가 원하는 ```ytInitialPlayerResponse = { 내용....};``` 을 찾았다면 이제 ```start_index``` 변수는 ```end_index``` 변수값보다 작아야 정상이고 이제 ```watch_html``` 변수에서 해당 값을 슬라이스 해야 합니다. ```start_index``` 변수에는 ```ytInitialPlayerResponse =``` 부터 시작하는 위치 값이 담겨 있기 때문에 우리는 ```ytInitialPlayerResponse =``` 를 제외 하기 위해서 ```start_index + len(start_str)``` 을 시작으로 ```end_index + 1``` 까지 슬라이스 해야 우리가 필요로 하는 ```{...내용...}``` 만 슬라이스 할 수 있습니다. 이렇게 슬라이스된 내용을 ```_json = json.loads(h)``` 하여 파이썬의 JSON 형태로 변환합니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
pprint.pprint(_json)
with open('output.txt', 'wt') as out:
    pprint.pprint(_json, stream=out)
</pre>
실제 ```_json``` 변수에는 상당히 많은 데이터가 있고 이를 그냥 ```print()``` 를 하면 알아보기가 불편하기 때문에 우리는 가독성을 좋게하여 분석을 좀 더 편하게 하기 위해 ```pprint``` 라이브러리를 사용해서 들여쓰기 상태로 화면에 출력하고 파일로도 저장합니다.
