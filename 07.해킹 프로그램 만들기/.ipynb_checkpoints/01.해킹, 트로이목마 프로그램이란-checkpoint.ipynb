# 해킹에 대한 이해

보통 일반적인 사람들이 이해하는 해킹은 누군가 시스템에 무단으로 침투하여 자료를 획득하거나 시스템을 무력화 시키거나 하는 부정적인 의미로 사용되기도 합니다만 좀 더 명확하게 이야기하자면 이렇게 시스템에 접속하여 권한을 획득하는 행위 까지를 해킹이라고 본다면 접속 후 무언가 파괴적인 행동을 하는것은 크래킹이라고 볼 수 있습니다. 그러나 현재 이런 구분 없이 유사적인 모든것을 총칭해서 해킹한다고 표현하는게 일반적입니다.

해킹의 방식에는 시스템 자체의 취약점을 이용하는 방법도 있겠지만 주로 악성코드를 배포하여 침투하는 방식이 훨씬 더 많이 사용됩니다. 오늘날의 악성코드에는 수십가지 이상의 종류가 있으며 그중 해킹에 사용되는 가장 고전적이며 강력한 트로이목마 방식에 대해서 알아보도록 하겠습니다.


## 악성코드

악성코드(Malware)는 사용자가 의도하지 않게 설치된 악의적인 목적을 갖는 모든 프로그램 혹은 코드를 총칭하며 우리가 알고 있는 컴퓨터 바이러스, 백도어, 웜, 애드웨어 등등을 포함하는 포괄적으로 사용되는 용어 입니다.


## 악성코드의 종류
사실 "악성코드" 라는 단어는 "컴퓨터 바이러스" 같은 단어에 비해 늦은 2000년대 들어서 나타난 단어입니다. 컴퓨터에 악영향을 끼칠 수 있는 이런 프로그램의 종류나 방식 그리고 감염 방법 등이 다양해지며 이를 분류하기 시작하였으며 이러면서 이렇게 많은 악성적인 프로그램을 총칭하는 단어로 사용되게 됩니다. 

![악성코드 종류](images/1.jpg)

#### - 컴퓨터 바이러스(Computer Virus)
컴퓨터 바이러스는 "바이러스" 처럼 행동하는 코드를 말합니다. 정상적인 "코드" 일부분에 자기 코드를 삽입하여 정상적인 코드가 실행될때 자신이 끼워놓은 코드가 실행되게 동작하며 자기 스스로는 실행되지 않습니다. 기존의 코드에 새로운 코드를 삽입하여 기존의 코드가 실행될때 삽입된 코드가 실행되게 만드는 방식은 그 자체로도 난이도가 높은 개발 방식으로 전체 악성코드중에 10% 도 되지 않지만 막상 감염이 되면 시스템이 받는 타격은 다른 악성코드에 비해 높은 편 입니다. 또한 원래 정상적인 프로그램에 코드를 삽입하기 위해선 정상적인 프로그램을 수정하게 되므로 백신에 탐지될 확률이 높아집니다.

#### - 웜(Worm)
웜의 강력한 특성은 네트워크를 통한 자가 복제 기능 입니다. 바이러스 처럼 개별 컴퓨터에서 감염되고 끝나는것이 아니라 네트워크 망을 통해 구현되기 때문에 가장 파괴적으로 동작하는 악성코드 입니다. 웜이 감염되면 네트워크 트래픽이 증가하는 증상을 보이게 됩니다. 웜 계열의 악성코드는 드라이버, 스크립트, 압축파일, 공유폴더, LAN, USB 등등에 자체 복제하여 확산되는 기능 등이 있습니다.

#### - 스파이웨어(Spyware)
개인정보 탈취를 목적으로 사용자의 동의 없이 설치되는 악성코드의 종류 입니다. 여기서 개인정보에는 인터넷 접속기록, 사용자가 입력한 아이디, 비밀번호 같은 키 입력 정보, 각종 시스템 로그 등을 수집하여 유출합니다. 대게 스파이웨어는 독립적인 실행파일 형태로 배포되거나 트로이잔 형태로도 배포됩니다.

#### - 루트킷(Rootkit)
루트킷은 원래 시스템의 Root 권한을 탈취할 목적의 프로그램등을 일컫는 용어로 사용되었으나 현재의 주된 목적은 시스템 침입 혹은 로그 등의 흔적을 탐지하지 못하도록 하는데 많이 사용됩니다. 루트킷이 그런 흔적을 원천적으로 제거하나 로그가 남지 않도록 하기 위해서는 일반적인 악성코드와 다르게 시스템의 관리자 이상의 권한을 획득할 수 있는 커널 모드로의 접근을 하는게 특징입니다. 윈도우 운영체제의 경우에는 실행위치에 따라 커널모드, 유저모드 루트킷이 있으며 커널모드의 루트킷을 제작하려면 NDIS(Network Driver Interface Specification), WDM(Windows Driver Model) 과 같은 시스템 드라이버 제작 기술을 알아야 할 정도로 난이도가 높은 측에 속하는 기술이며 이정도의 루트킷은 감염되면 치료하기도 어렵고 더 파괴적인 루트킷 같은 경우에는 하드웨어 자체를 망가트리는 악성코드도 있습니다.

#### - 익스플로잇(Exploit)
익스플로잇은 사실 악성코드 자체를 지칭하기보다는 악성코드가 침투하는 방식을 말하는 용어 입니다. 뉴스 같은데 자주 나오는 "취약점 공격" 이라고 말하는 것을 익스플로잇 이라고 합니다. 익스플로잇은 컴퓨터 소프트웨어의 자체 결함 혹은 취약점을 악용하여 시스템의 권한을 획득하거나 공격하는 행위를 말하며 다른 악성코드와는 다르게 시스템의 파일 형태로 감염되거나 전염되는 것과 달리 파일이 존재하지 않거나 간단한 스크립트 문장만으로도 공격이 가능한 경우도 있습니다. [Exploit-DB](https://www.exploit-db.com/) 에서는 일반적으로 보안전문가나 화이트 해커를 지망하는 사람들은 이런 일련의 익스플로잇 방법에 대해 토론하기도 합니다.

#### - 애드웨어(Adware)
사용자가 동의하지 않는 광고에 노출 시키는게 목적인 악성코드 입니다. 지금은 점점 없어지고 있지만 마이크로소프트의 인터넷 익스플로어의 Active X, BHO 등의 기술 형태로 가장 많이 배포되었습니다.

```python
# 애드웨어 샘플코드
import tkinter as tk
from PIL import Image, ImageTk
import os
import requests


def close_tk(event):
    root.quit()


url = "https://cdn.inflearn.com/public/courses/324037/course_cover/626bed53-f79c-4fbf-b513-b18772f0ef6c/%E1%84%82%E1%85%A1%E1%86%B7%E1%84%87%E1%85%A1%E1%86%A8%E1%84%89%E1%85%A1_100.png"
cur_path = os.path.dirname(os.path.realpath(__file__))
root = tk.Tk()
root.overrideredirect(True)
image = Image.open(requests.get(url, stream=True).raw)
img_width, img_height = image.size

tk_image = ImageTk.PhotoImage(image)
label = tk.Label(image=tk_image)
label.grid(row=1, column=0)
label.bind("<Button-1>", close_tk)

screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()

x_coord = int(screen_width - img_width)
y_coord = int(screen_height - img_height)

root.geometry("{}x{}+{}+{}".format(img_width, img_height, x_coord, y_coord))
root.mainloop()
```

#### - 랜섬웨어(Ransomware)
랜섬웨어는 사실 보안 프로그램의 파일 보안기능과 유사한 기능을 역으로 공격에 사용한 방식으로 컴퓨터에 있는 파일들을 암호화 시켜 사용자가 접근할 수 없게 한 후 돈을 요구하는 형태의 악성코드 입니다. 랜섬웨어에서는 양방향 암호화 혹은 단방향 암호화 알고리즘을 사용하는데 단방향 암호화 같은 경우에는 암호화 되기의 원래 시점으로 돌아갈 수 있는 방법 자체가 없으므로 감염시 시스템 복구가 불가능하게 됩니다.

```python
# 랜섬웨어 샘플 코드
from cryptography.fernet import Fernet
import os

cur_path = os.path.dirname(os.path.realpath(__file__))

KEY_FILENAME = "{}\\mykey.key".format(cur_path)
SOURCE_FILE = "{}\\test_src.txt".format(cur_path)
ENCRYT_FILE = "{}\\test_enc.txt".format(cur_path)
DECRYPT_FILE = "{}\\test_dec.txt".format(cur_path)


def write_key():
    key = Fernet.generate_key()
    with open(KEY_FILENAME, "wb") as f:
        f.write(key)


def load_key():
    return open(KEY_FILENAME, "rb").read()

ㅠ
def encrypt_file(source_filename, target_filename, key):
    f = Fernet(key)
    with open(source_filename, "rb") as file:
        f_data = file.read()
    e_data = f.encrypt(f_data)
    with open(target_filename, "wb") as file:
        file.write(e_data)


def decrypt_file(source_filename, target_filename, key):
    f = Fernet(key)
    with open(source_filename, "rb") as file:
        e_data = file.read()
    f_data = f.decrypt(e_data)
    with open(target_filename, "wb") as file:
        file.write(f_data)


if __name__ == "__main__":
    # 키 생성
    if not os.path.exists(KEY_FILENAME):
        write_key()
    # 키 로드
    key = load_key()
    # 암호화
    encrypt_file(SOURCE_FILE, ENCRYT_FILE, key)
    # 복호화
    decrypt_file(ENCRYT_FILE, DECRYPT_FILE, key)
```
