# 프로그램 배포

파이썬으로 작성된 프로그램은 스스로 실행할 수 없고 반드시 python.exe 에 의해 .py 파일을 실행시키는 구조로 동작합니다. 이는 컴파일되어 자체적으로 실행 기능을 갖는 프로그램과 달리 인터프리터 언어의 특징이기도 합니다. 지금까지 작성한 백도어를 실행하기 위해 사용자에게 파이썬을 설치하라고 얘기할 수는 없습니다. 그래서 지금까지 파이썬으로 작성한 .py 파일을 윈도우에서 실행가능한 .exe 파일로 변환하는 작업을 거쳐야 합니다.

### pyinstaller 4.3

> pip install pyinstaller --upgrade

원칙적으로 파이썬으로 작성된 ```.py``` 을 실행하기 위해서는 사용자의 컴퓨터에 파이썬이 설치되어 있어야 하고 해당 프로그램이 사용하는 모든 라이브러리들도 설치되어있어야 됩니다. 

pyinstaller 는 이렇게 작성된 .py 파이썬 프로그램과 프로그램에 사용된 종속되는 모든 라이브러리들을 하나의 패키지로 묶어 실행가능한 상태로 패킹하는 툴 입니다. 사용자의 PC에 파이썬이 설치되지 않았어도 이렇게 pyinstaller 에 의해 패킹된 프로그램을 실행할 수 있습니다. 이렇게 pyinstaller 는 ```.py``` 파일을 실행가능하게 만드는 목적 외에도 ```.py``` 파일은 그 자체로 소스코드를 확인할 수 있기 때문에 이를 방지하기 위한 목적도 있습니다. <font color="red" style="background-color:#eeeeee">단지 .py 파일을 1차적으로 변환하는 정도이지 완벽하게 소스코드가 보안되게 하는것은 아닙니다.</font>

또한 pyinstaller 는 윈도우즈 운영체제 뿐 아니라 Mac OS X, 리눅스를 지원합니다. 단, 만약 32비트 윈도우에서 실행가능한 프로그램을 작성하려면 pyinstaller 를 32비트 윈도우에서 실행해야 하고 64비트 윈도우에서 실행가능하게 하려면 64비트 윈도우환경에서 pyinstaller를 실행해야 합니다. 그리고 리눅스에서 실행가능한 프로그램을 작성하려면 pyinstaller 가 리눅스에서 실행되어야 합니다. 해당 강좌에서는 윈도우용으로 진행합니다.


### 사용법

> pyinstaller [옵션] sample.py

![pyinstaller](images/15.jpg)

pyinstaller 를 실행하면 최종적으로 위의 이미지에서 처럼 dist, build 폴더와 .spec 파일이 생성됩니다. ```.spec``` 파일은 최초 pyinstaller 가 실행되면 현재 실행된 정보가 기록되는 파일인데 이 파일을 수정하여 차후에 해당 프로젝트를 pyinstaller 로 리빌드 할때 좀 더 자세한 환경 설정을 할 수 있습니다. build 폴더에는 pyinstaller 에 필요한 작업 파일들이 생성되고 최종 결과는 dist 폴더에 저장됩니다. [[spec file 정보 공식링크]](https://pyinstaller.readthedocs.io/en/stable/spec-files.html?highlight=block_cipher#spec-file-operation)

### 해킹툴 패킹하기

지금까지 우리가 작성한 해킹툴을 pyinstaller 를 이용하여 단독 실행이 가능한 exe 로 변경 해보도록 하겠습니다. 변경시에 우리는 [[아이콘]](https://icon-icons.com/ko/download/14823/ICO/256/) 을 활용할 예정입니다. 가정 먼저 할 일은 지금까지 작성한 프로그램이 ```.py``` 을 실행했을때 오류가 없는지를 먼저 확인해야 합니다.

> pyinstaller -F -w -i notepad.ico --key testkey -n pypad --clean remote_client.py

![pyinstaller](images/18.jpg)

위의 이미지에서 처럼 실행을 하고 나면 ./dist 폴더에 pypad.exe 라는 파일이 i 옵션으로 설정된 아이콘의 모양으로 생성된것을 확인할 수 있습니다. 이렇게 생성된 파일은 <font color="red" style="background-color:#e9e9e9">최초 명령프롬프트 상에서 실행을 해보고 반드시 최종적으로 오류가 없는지를 확인해야 합니다.</font>

![pyinstaller](images/20.jpg)

윈도우에서 문제 없는 똑같은 파일을 우분투에서 pyinstaller 로 작업했을 경우 위의 이미지에서 처럼 Import 오류가 발생하는 경우도 있습니다. 이런경우에 --hidden-import 을 사용하여 처리 해야 합니다.

![pyinstaller](images/21.jpg)

물론 리눅스는 윈도우처럼 더블클릭만으로 실행파일을 실행할 수 있는게 아니기 때문에 우리가 작성한 해킹툴 같은 프로그램은 윈도우에서보다 활용성은 떨어지는게 사실입니다만 해킹툴이 아닌 다른 파이썬 프로그램을 작성하여 다른 환경에 배포하게 되는경우에는 유용하게 사용될 수 있는 부분입니다.


### .spec 파일

위에서 우리가 pyinstller 를 실행시킬때 여러가지 옵션을 사용했습니다만 옵션이 많은 경우 매번 pyinstaller 를 실행할 때마다 이를 작성한다는건 비효율적인 내용입니다. 이런경우 ```.spec``` 파일을 활용하면 좀 더 편리하게 사용할 수 있습니다.

````python
%%html

# -*- mode: python ; coding: utf-8 -*-

block_cipher = pyi_crypto.PyiBlockCipher(key='abcd')

a = Analysis(['remote_client.py'],
             pathex=['/home/vmware/Python/PyTrojan'],
             binaries=[],
             datas=[],
             hiddenimports=['pynput.keyboard._xorg', 'pynput.mouse._xorg'],
             hookspath=[],
             runtime_hooks=[],
             excludes=[],
             win_no_prefer_redirects=False,
             win_private_assemblies=False,
             cipher=block_cipher,
             noarchive=False)
pyz = PYZ(a.pure, a.zipped_data,
             cipher=block_cipher)
exe = EXE(pyz,
          a.scripts,
          a.binaries,
          a.zipfiles,
          a.datas,
          [],
          name='pypad',
          debug=False,
          bootloader_ignore_signals=False,
          strip=False,
          upx=True,
          upx_exclude=[],
          runtime_tmpdir=None,
          console=False , icon='notepad.ico')
````

위의 ```.spec``` 파일을 에디터 프로그램으로 열어보면 우리가 실행시에 작성한 대부분의 옵션이 이미 기록되어있는 것을 확인 할 수 있습니다. 이제부터는 직접 옵션을 주어서 실행하지 않고 수정될 내용이 있으면 ```.spec``` 파일을 수정하면 됩니다. 이렇게 수정된 ```.spec``` 파일의 내용대로 pyinstaller 를 다시 빌드 하려면

> pyinstaller pypad.spec

위처럼 pyinstaller를 ```.py``` 파일이 아닌 ```.spec``` 파일로 실행하면 됩니다.


### ResourceHacker [[공식사이트]](http://www.angusj.com/resourcehacker/)

pyinstaller 에서 제공하는 ```-i``` 옵션으로 아이콘을 바꾸는 방법외에도 외부 프로그램인 Resource Hacker 같은 프로그램으로 현재 이미 작성 완료된 exe 파일의 icon 을 다른 ico, bmp 혹은 다른 exe 프로그램이나 dll 파일 내의 아이콘으로 변경할 수 있습니다.


### 파일병합

해킹 프로그램을 감추기 위해 또 많이 사용되는 방법중 한가지가 2가지의 파일을 하나로 합치는 파일 병합 기법입니다. 파일 병합은 외부의 툴을 이용하여 A 프로그램과 B 프로그램을 합쳐서 새로운 C 프로그램을 만들어내는 기법이 있고 아니면 A 프로그램에 B 프로그램을 삽입하는 PE Injection 기법이 있습니다. PE 라는 것은 Portable Executable 의 약자로 윈도우 운영체제에서 사용되는 실행파일, DLL 파일등의 구조를 말하는데 말 그대로 A 라는 프로그램의 구조를 분석하여 A 파일 자체에 B 파일을 강제로 삽입하고 실행시킬수 있는 기술입니다. 보통 이렇게 실행되는 악성코드를 Dropper 라는 이름으로 분류하기도 합니다. 물론 PE 파일의 구조를 알아야 하기 때문에 기술적으로 손쉽게 구현될 수는 없고 현재 많은 백신 프로그램이 이렇게 PE Injection 된 파일을 휴리스택 검출 방식으로 쉽게 잡아내고 있습니다. 파이썬으로 PE Injection 를 구현한 부분에 대한 내용은 [[링크]](https://axcheron.github.io/code-injection-with-python/) 에서 볼 수 있습니다. 단, 해당 블로그의 코드는 현재 정상적으로 실행되지 않으니 보안에 관심있는 분들만 이론적으로 보시기엔 훌륭한 글이라 생각됩니다.


### 프로세스 목록

물론 어떻게든 실행된 프로그램은 윈도우의 경우 작업관리자의 프로세스 목록에 표기가 되며 사용자가 강제종료 할 수 있습니다. 물론 대부분의 악성코드는 사용자가 예측하기 힘든 System.exe, update.exe 등의 애매한 실행 파일명으로 위장을 하는 경우가 대부분입니다만 더 강력한 악성코드는 이렇게 사용자에게 강제종료 당하지 않게 하기 위해 실행파일 형태가 아닌 서비스로 동작한다던지 아니면 스케쥴러 같은데 등록하여 주기적으로 본체는 숨기고 스스로를 랜덤한 파일명과 프로세스 명으로 복제하여 실행시키기도 하며 아예 작업관리자의 프로세스 목록에서 스스로를 감춰버리기도 합니다. 이 [[링크]](https://777bareman777.github.io/2020/01/16/WindowsAPIHookingStealthProcess-1/) 는 보안에 관심있는 분들은 한번쯤 읽어볼만한 내용인것 같습니다.
