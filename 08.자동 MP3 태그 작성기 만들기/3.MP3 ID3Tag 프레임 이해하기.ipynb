### 4.2.1 Text information frames

위에 가장 많이 등장하는 4.2.1 유형은 text information frames 유형으로 단어 그대로 텍스트를 정보를 포함하는 프레임 구조 입니다. 텍스트 정보 프레임의 아이디는 보통 "T" 로 시작합니다. <font color="red" style="background-color:#eeeeee;">Text information frame 에서 실제 데이터의 첫 바이트는 텍스트 인코딩을 표기</font>합니다. 텍스트 데이터는 0x00 으로 종료되고 0x00 이후의 문자는 무시됩니다. 텍스트 인코딩은 ID3v2.3 에서는 0x00(아스키), 0x01(유니코드) 로만 구분되며 v2.4 에서는 좀 더 세밀하게 설정됩니다.


![img](images/5.jpg)

위의 이미지에서 처럼 hex 뷰어로 본 MP3 파일을 살펴보면 ```TABL``` 로 시작하는 프레임 헤더인 A길이(10바이트) 의 데이터를 보면 ```54, 41, 4C, 42, 00, 00, 00, 29, 00, 00``` 인것을 확인할 수 있습니다. 위의 frame id 항목을 보면 ```TALB``` 는 앨범 이름이라는것을 알 수 있습니다. 위에서 얘기한 프레임 헤더 구조를 적용해보면 ```54, 41, 4C, 42, 00, 00, 00, 29, 00, 00``` 데이터에서 첫 4바이트 ```54, 41, 4C, 42``` 는 ```TALB``` 라는 글자의 값이고 그 뒤에 ```00, 00, 00, 29``` 는 사이즈 정보, ```00, 00``` 는 플래그 값인걸 알 수 있습니다. 

![img](images/6.jpg)

여기서 사이즈 정보인 ```00, 00, 00, 29``` 를 10진수로 변환해보면 ```41``` 이란것을 알 수 있고 실제 데이터는 헤더 이후의 41바이트까지에 담긴것을 알 수 있습니다. 위의 이미지에서 실제 데이터를 보면 첫 바이트가 ```01``` 인것을 확인할 수 있는데 이는 실제 데이터가 인코딩 되었다는 의미가 되고 한글로 출력하기 위해선 디코딩이 필요하다고 볼 수 있습니다. <font color="red" style="background-color:#eeeeee;">기본적으로 ID3v2.3 에서는 UTF-16 을 사용합니다.</font> 따라서 위의 이미지에서 실제 데이터는 ```01``` 을 제외한 ```FF, FE, 52....``` 부터가 실제 앨범 제목 데이터가 됩니다.

````python
from io import BytesIO

filename = "bensound-newdawn.mp3"

def decode_size(str_hex):
    slice_hex = [str_hex[i:i+2] for i in range(0, len(str_hex), 2)]
    results = ""
    for h in slice_hex:
        results += "{0:b}".format(int(h, 16)).zfill(7)
    return int(results, 2)

mp3 = open(filename, "rb")
header_size = 10
header_data = mp3.read(header_size)
id3_header = BytesIO(header_data)

ID3_TAG = id3_header.read(3)
ID3_VERSION = id3_header.read(2)
ID3_FLAGS = id3_header.read(1)
ID3_SIZE = id3_header.read(4)
ID3_BYTES = decode_size(ID3_SIZE.hex())

print(ID3_TAG, ID3_VERSION, ID3_FLAGS, ID3_SIZE, ID3_BYTES)

while True:
    try:
        temp_bytes = BytesIO(mp3.read(10))
        fid = temp_bytes.read(4).decode()
        size = int(temp_bytes.read(4).hex(), 16)
        flags = temp_bytes.read(2)
        if fid == "TALB":
            vio = BytesIO(mp3.read(size))
            text_encoding = vio.read(1)
            text_value = vio.read()
            text_value.rstrip(b'\x00')
            print(fid, size, text_value.decode("UTF-16"))
            break
    except:
        break
````

위의 코드는 MP3 파일안에서 TALB 프레임을 구하고 해당 프레임의 데이터값을 출력하는 내용입니다. 임시로 작성한 코드기 때문에 TALB 를 찾으면 break 문으로 while 문을 탈출하게 되어있습니다. 위 코드를 좀 더 자세하게 살펴보도록 하겠습니다.

<br><pre style="background-color:#eeeeee;margin:0px;padding:10px;">

</pre>

위의 코드에서 ```mp3``` 변수는 위에서 open 한 MP3 파일의 오브젝트인데 헤더크기(10바이트)를 읽어 BytesIO 형태로 변환해서 사용합니다. BytesIO 형태로 변환하여 사용하는 이유는 실제 파일을 열였을때와 마찬가지로 BytesIO 역시 파일포인터가 동작가능하기 때문인데 파일 포인터라는개념에 대해 조금 이해하고 넘어갈 필요가 있습니다.

    text_encoding = vio.read(1)
    text_value = vio.read()
    text_value.rstrip(b'\x00')
    if text_encoding == b"\x01":
        print(fid, size, text_value.decode("UTF-16"))
    else:
        print(fid, size, text_value.decode("cp949"))
</pre>
여기서 ```fid``` 값이 "TALB" 인 경우에 다시 mp3 파일을 방금 프레임 헤더에서 구한```size``` 만큼 읽어 (mp3 변수의 파일포인터는 다시 그만큼 이동) BytesIO 형태의 ```vio``` 변수에 저장합니다. 이제 다시 ```vio``` 변수를 1바이트 만큼 읽어서 인코딩 정보를 추출하고 실제 데이터는 ```text_value``` 변수에 저장합니다. 이때 ```rstrip()``` 함수를 사용하여 공식문서에 표기된 데로 ```\x00```(null) 문자 뒤는 제거합니다. 그리고 인코딩 여부에 따라 인코딩이 된 값이면 UTF-16 으로 디코딩하고 그렇지 않다면 한글 윈도우의 경우 시스템 기본 인코딩 방식인 cp949 로 인코딩 하여 출력합니다.

위의 이미지에서 처럼 실제 데이터중 앞의 바이트만 살펴보면 ```01 6B 6F 72 FF FE 00 00 FF FE....``` 로 시작하는것을 알 수 있습니다. 위의 표와 대입해서 분석해보면 첫 바이트는 인코딩 여부의 판단인데 ```01``` 값이니 해당 데이터는 인코딩된 데이터로 차후 디코딩을 해야한다는것을 알수 있고 그 뒤에 3바이트 ```6B 6F 72``` 는 언어를 표기한건데 우측 decode text 를 살펴보면 ```kor``` 이 눈으로 보이는것을 확인 할 수 있습니다.

````python
from io import BytesIO

filename = "bensound-newdawn.mp3"

def decode_size(str_hex):
    n = 2
    slice_hex = [str_hex[i:i+n] for i in range(0, len(str_hex), n)]
    results = ""
    for h in slice_hex:
        results += "{0:b}".format(int(h, 16)).zfill(7)
    return int(results, 2)

mp3 = open(filename, "rb")
header_size = 10
header_data = mp3.read(header_size)
id3_header = BytesIO(header_data)

ID3_TAG = id3_header.read(3)
ID3_VERSION = id3_header.read(2)
ID3_FLAGS = id3_header.read(1)
ID3_SIZE = id3_header.read(4)
ID3_BYTES = decode_size(ID3_SIZE.hex())

print(ID3_TAG, ID3_VERSION, ID3_FLAGS, ID3_SIZE, ID3_BYTES)

ID3_DATA = BytesIO(mp3.read(ID3_BYTES))

while True:
    temp_bytes = BytesIO(ID3_DATA.read(10))
    if ID3_DATA.tell() == ID3_BYTES or temp_bytes == b"":
        break
    try:
        fid = temp_bytes.read(4).decode("UTF-8")
        size = int(temp_bytes.read(4).hex(), 16)
        flags = temp_bytes.read(2)
        if fid == "TALB":
            vio = BytesIO(ID3_DATA.read(size))
            text_encoding = vio.read(1)
            text_value = vio.read()
            text_value.rstrip(b'\x00')
            if text_encoding == b"\x01":
                print(fid, size, text_value.decode("UTF-16"))
            else:
                print(fid, size, text_value.decode("cp949"))
        elif fid == "USLT":
            vio = BytesIO(ID3_DATA.read(size))
            text_encoding = vio.read(1)
            text_language = vio.read(3)
            if text_encoding == b"\x01":
                text_value = vio.read().decode("UTF-16")
            else:
                text_value = vio.read().decode("cp949")
            print("{:*^50}".format(" 노래가사 "))
            print(text_value)
    except Exception as e:
        pass
````

위 코드는 ID3v2.3 의 가사 프레임을 읽어오는 기능을 추가한 코드 입니다. 기존에 작성한 코드를 조금 수정하였습니다. 수정된 부분을 좀 더 자세히 살펴보면..

</pre>
이제 MP3 파일 전체를 핸들링 하지 않고 ID3_BYTES 에 구한 ID3v2.3 태그의 전체 크기만큼만 읽어들여서 작업을 하려고 BytesIO 형태의 ```ID3_DATA``` 변수에 저장하였습니다. 이제 MP3 파일을 직접 오픈한 ```mp3``` 변수는 사용하지 않습니다.
