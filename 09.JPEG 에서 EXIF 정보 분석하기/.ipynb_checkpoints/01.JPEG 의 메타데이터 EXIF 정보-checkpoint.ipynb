우리가 디지탈카메라 등에서 흔하게 사용하는 확장자가 **.jpg, .jpeg, .jpe** 파일인 이미지 포맷은 정지 화상을 위해 만들어진 손실 압축 방법 표준의 형식입니다. JPEG은 이미지를 어떻게 압축하는가에 대한 방식을 말하고 이 JPEG 로 압축된 데이터를 JFIF 형태로 저장하면 우리가 말하는 .jpg, .jpeg 파일이 되는거라고 볼 수 있습니다.

위 표는 [[위키피아의 JPEG 구조표]](https://en.wikipedia.org/wiki/JPEG#Syntax_and_structure) 내용을 재작성한 내용입니다. 우리는 평상시 jpg 파일 하면 그냥 이미지를 떠올렸습니다만 실제 jpg 파일은 위의 표에서처럼 우리가 평상시 들어본적 없는 용어들의 이름으로 복잡하게 구성되어 있으며 jpg 뷰어가 이런 데이터를 분석하고 디코딩하여 우리에게 이미지를 보여주게 됩니다.

````python
MARKER = {
    b"\xFF\xD8": "Start of Image",
    b"\xFF\xE0": "Application Segment 0",
    b"\xFF\xE1": "Application Segment 1",
    b"\xFF\xDB": "Quantization Table",
    b"\xFF\xC0": "Start of Frame",
    b"\xFF\xCC": "Define Arithmetic Coding",
    b"\xFF\xC4": "Define Huffman Table",
    b"\xFF\xC8": "JPEG Extensions",
    b"\xFF\xD0": "Restart Marker",
    b"\xFF\xDA": "Start of Scan",
    b"\xFF\xDB": "Define Define Quantization Table",
    b"\xFF\xDC": "Define Number of Lines",
    b"\xFF\xDD": "Define Restart Inteval",
    b"\xFF\xD9": "End if Image",
}

filename = "note9.jpg"
with open(filename, 'rb') as f:
    data = f.read()

    for k, v in MARKER.items():
        offset = data.find(k)
        if offset >= 0:
            print("Key:{}, Name:{}, Offset:{}".format(k, v, offset))
````

위 코드는 간단하게 강좌에서 사용되는 샘플이미지인 ```note9.jpg``` 파일 내부 세그먼트들의 마커 정보를 출력하는 예제 코드 입니다. JPEG 는 이렇게 여러가지의 세그먼트라 불리는 구조로 이루어져 있습니다. 이 강좌에선 우리는 JPEG 파일의 전체 구조를 분석하고 이해하려는 목적은 아니고 위의 **JPEG 구조표** 에서 붉은 색으로 표시된 Application Segment 1 인 EXIF 정보에 대해서만 다룰 예정입니다. 

### EXIF (<font color="red">EX</font>changeable <font color="red">I</font>mage <font color="red">F</font>ile format)

디지털 카메라 등에서 사용되는 이미지 파일 메타데이터 포맷으로 가장 중요한 특징은 카메라의 기종, 촬영시간, GPS 정보, 초점거리, 플래시 발광여부 등등등 수많은 메타데이터를 기록할 수 있다는 점 입니다. 이런 EXIF 정보는 꼭 JPEG 파일만 사용되는것은 아니고 TIFF 6.0, RIFF, WAV 파일 그리고 PNG (v1.5 버전이상) 등 여러 포맷의 일부로 저장되는 메타데이터 포맷입니다.

![image](images/exif.jpg)

이미지에 EXIF 정보가 존재한다면 이미지 뷰어 프로그램을 사용하여도 EXIF 정보를 볼 수 있지만 위 처럼 윈도우 운영체제의 탐색기에서 이미지 파일을 우클릭하여 속성만 살펴보아도 간단하게 EXIF 정보를 살펴볼 수 있습니다. EXIF 정보는 사진에 대한 추가 정보를 많이 담고 있기 때문에 이미지 외에도 여러가지 데이터를 동시 기록한다는 장점도 있지만 GPS 정보같이 민감한 데이터는 개인 정보 유출 혹은 사생활 침해도 될 수 있기 때문에 좀 더 각별하게 신경을 써야 하는 부분도 있습니다. 참고로 대부분의 스마트폰 카메라의 GPS 옵션은 디폴트 값이 False 입니다. 따라서 사진에 GPS 정보를 기록하려면 카메라의 설정에서 위치정보 옵션을 설정해야 합니다.

![image](images/jpeg_structure.jpg)

EXIF 포맷은 일본전자정보 기술산업협회(JEITA)에 의해 개발되었는데 위 이미지는 [[JEITA Exif 2.2 문서]](http://exif.org/Exif2-2.PDF)에서 어플리케이션 마커(APP1) 일부의 내용을 재작성한 표 입니다. 강좌 처음의 표에서 APP<em style="font-weight:bold;">n</em> 으로 기술된 부분입니다.

어떻게 보면 정말 잘 설명해놓은 표 이미지지만 태어나서 처음 보는 입장에선 무슨 얘긴지 도통 알수가 없습니다. 문제는 우리가 필요로 하는 EXIF 데이터는 위의 이미지에서 처럼 일반적으로 APP1 안에 기록되는건 알겠는데 APP1도 단순하게 생긴거 같지가 않습니다. 어떻게 저 APP1 에 접근하고 또 정보를 얻어와야 하는지 아무것도 모르니 일단 16진수 뷰어/편집기 를 통해 JPEG 파일부터 열여보는걸로 시작해보도록 하겠습니다. 16진수 뷰어는 이전 MP3 강좌에서도 사용했지만 각자의 운영체제에 맞는 프로그램을 찾으셔야 하고 윈도우 운영체제 같은경우 [[Hxd]](https://mh-nexus.de/en/downloads.php?product=HxD20) 를 추천합니다.


![image](images/1.jpg)

위의 이미지는 일반적인 스마트폰으로 촬영한 jpg 파일을 16진수 뷰어로 본 모습 입니다. 위의 [JPEG 구조표](#JPEG-구조)에서 처럼 JPEG 파일의 시작 (SOI)은 항상 0xFF, 0xD8 로 시작하는것을 볼 수 있습니다. 그리고 우측의 Decode Text 쪽을 보면 얼핏 우리가 알아볼 수 있는 samsung 등의 알파벳 글자도 보이는것도 알 수 있습니다.

![image](images/2.jpg)

JPEG 이미지의 시작임을 알리는 0xFF, 0xD8 바로 뒤에는 0xFF, 0xE1 값이 바로 자리 하고 있습니다. 이 부분이 위의 [JPEG 구조표](#JPEG-구조)에서 빨간색 박스로 강조해놓은 App<em style="font-weight:bold;">n</em> 영역의 마커 표시인데 0xFF, 0xE1 은 APP1 의 마커이고 우리가 필요로하는 EXIF 정보가 이곳에 담겨 있습니다. <font color="red" style="background-color:#eeeeee;font-weight:bold;">여기서 간과해서 안되는 부분은 항상 위의 상황처럼 0xFF, 0xD8 바로 뒤에 APP1 의 값이 존재한다는것은 아닙니다.</font> 만약 이미지에 JFIF 정보가 있다면 JFIF 정보는 APP0 을 사용하기 때문에 APP1 보다 먼저 등장할 수도 있습니다.

![image](images/4.jpg)

APP1 마커 뒤에는 사실 APP1 의 데이터 크기가 표기됩니다만 위와 같이 그냥 최대값 0xFF, 0xFF 로 기재되는 경우가 있습니다. 그렇기 때문에 이 값을 실제 사용하기에는 무리가 있을 수 있습니다.

![image](images/3.jpg)

EXIF 헤더는 위의 이미지에서 처럼 "Exif" 문자열을 나타내는 16진수 값인 0x45(E), 0x78(x), 0x69(i), 0x66(f) 그리고 Null 문자 2바이트 0x00, 0x00 로 고정값이라보 보면 됩니다. 

![image](images/5.jpg)

Exif 헤더 바로 뒤에는 8바이트 TIFF 헤더가 위치합니다. 앞에서부터 살펴보면 앞의 2바이트는 데이터의 바이트 정렬을 정의합니다. 위의 이미지에서는 0x49, 0x49 는 아스키문자로 II 로 표기되는데 "Intel" 유형의 바이트 정렬을 의미하고 0x4D, 0x4D 는 아스키문자로 MM 으로 표기되는데 이는 "Motorola" 유형의 바이트 정렬을 의미합니다. 인텔방식은 리틀엔디안, 모토로라 방식은 빅엔디안을 사용 하는데 쉽게 리틀엔디안은 작은수를 오른쪽부터, 빅엔디안은 큰수를 오른쪽부터 기입하는 방식이라고 보시면 됩니다. 예를 들어 0x123456 이라는 값을 표기할때 리틀엔디안(II) 은 0x56, 0x34, 0x12 로 기입하고 빅엔디안은 0x12, 0x34, 0x56 으로 기입하게 됩니다. <font color="red" style="background-color:#eeeeee;font-weight:bold;">이미지 파일내의 모든 EXIF 데이터를 분석할때 이 바이트 정렬 방식을 기준으로 하기 때문에 상당히 중요한 부분</font>입니다.

![image](images/6.jpg)

위의 이미지를 보면 방금전 이미지와 다르게 0x4D, 0x4D 로 시작하는 "MM" (빅엔디안 방식) 형식으로 모토로라 방식의 바이트 정렬의 샘플 이미지 입니다. 바이트정렬 다음 2바이트는 0x2A, 0x00 (인텔) 혹은 0x00, 0x2A(모토로라) 고정값 입니다. 그리고 마지막 4바이트는 첫번째 IFD 에 대한 오프셋인데 현재 TIFF 헤더값이 바이트오더 2바이트 + 마커 2바이트 + 오프셋 4바이트 = 총 8바이트로 작성되어있으니 첫번째 IFD 오프셋에 들어가는 값은 8바이트로 고정값이 됩니다. 여기서 오프셋이란 어떤 위치(주소)를 가르키는 용어인데 어떤 위치가 고정되어있는 한 지점(주소)을 말하는게 아니라 특정 기준 위치에서부터 얼만큼 떨어져 있다는 상대적 주소값을 말 합니다. 당장 이해가 안가더라도 계속 강좌를 진행하다 보면 이해를 하실 수 있으리라 생각됩니다.

### 파이썬으로 APP1 영역 구하기

````python
import struct

filename = "note9.jpg"
f = open(filename, "rb")
data = f.read(2)
if data[0:2] != b"\xFF\xD8":
    print("Not JPEG file!!")
    exit(0)

while True:
    H = f.read(2)
    if H == b"\xFF\xE1":
        data = f.read(16)
        length = struct.unpack(">H", data[0:2])[0]
        if data[2:6] != b"Exif":
            print("Not EXIF!")
            exit(0)
        segment_data = f.read(length - 16)
        exif = H + data + segment_data
        break
    elif H == b"":
        break
f.close()
print(exif[-10:])
````

위의 코드는 jpg 파일을 읽어서 APP1 영역을 찾고 분리하는 간단한 샘플 코드 입니다. 물론 위 코드는 예외처리가 되지 않은 샘플로 사용된 이미지에 최적화된 코드라 범용적으로 사용하기에는 무리가 있는 코드 입니다. 강좌의 특성상 최대한 이해를 돕기 위해 복잡한 예외처리는 하지 않고 진행할 예정입니다.

저장된 ```data[0:2]``` 변수의 사이즈 정보를 ```struct.unpack``` 함수의 포맷 문자열을 사용하여```>H``` (unsigned short) 형태로 변환하여 사이즈를 구하고 ```data[2:6]``` 에 "Exif" 문자열이 있는지 확인 한 후 간단한 오류 처리를 했습니다. ```f.read(length-16)``` APP1 크기만큼을 읽어 ```segment_data``` 변수에 저장합니다. 여기서도 ```segment_data``` 변수에 따로 APP1 만큼을 미리 읽어 저장하는 이유는 차후 IFD 정보에 접근하기 용이하게 하기 위함인데 샘플로 제공된 이미지 처럼 사이즈 정보가 단순 최대값 0xFF, 0xFF 로 기록되어있을지라도 파일 전체 크기만큼을 모두 분석할 필요는 없기 때문에 이렇게 하는게 더 효율적입니다. 현재 0xFF, 0xE1 마커를 비롯하여 ```segment_data``` 내용까지 ```exif``` 변수에 모두 저장합니다. 이렇게 되면 더이상 파일 객체를 사용하며 파일 포인터를 신경쓸 필요없이 ```exif``` 변수안에 저장된 데이터로 모든 분석을 할 수 있습니다.

### 바이트 변환
```struct.unpack()``` 함수를 사용해서 바이트 데이터를 원하는 우리가 원하는 형태로 변환하는 이유는 파이썬과 C언어 상의 자료 형태가 상이하기 때문이라고 볼 수 있습니다. 프로그래밍에서 표준처럼 사용되는 C/C++ 에서는 정수의 크기를 고려해서 프로그래밍을 하는데 만약 정수의 범위가 –2,147,483,648 ~ 2,147,483,647 범위 안이면 4바이트의 int 형을 쓰고 만약 2,147,483,647 를 넘어가되 음수는 필요없다 하면 ```unsigned int``` 형을 씁니다. 만약 -32,768 ~ 32,767 범위라면 2바이트의 ```short``` 형을 사용합니다. 파이썬에서는 임의 정밀도라는 개념으로 ```int``` 를 사용하며 ```int``` 는 시스템 가용한 모든 메모리를 사용하여 수를 저장할 수 있습니다. 그렇기 때문에 파이썬 환경에서의 ```int``` 형과 C/C++ 환경에서의 ```int``` 는 다른 메모리 크기를 갖습니다. 결론적으로 이렇게 표준 데이터 크기로 작성된 데이터를 읽고 쓰기 편하게 하기 위해서 파이썬에서는 ```struct``` 라이브러리를 사용하게 됩니다.
