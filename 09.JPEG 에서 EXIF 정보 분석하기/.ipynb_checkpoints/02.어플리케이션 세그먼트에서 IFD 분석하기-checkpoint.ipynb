실제 데이터 총 바이트 길이는 위 표의 "바이트/콤포넌트" x 컴포넌트 갯수 를 계산하여 얻을 수 있습니다. 이렇게 계산하여 얻어진 바이트 길이가 4바이트 미만이면 실제 데이터는 엔트리의 데이터/오프셋 영역에 작성되지만 4바이트를 넘어가게 되면 데이터/오프셋 영역에는 실제 데이터가 위치하는 오프셋 주소값만 기록되게 됩니다. 위의 표들이 간단하게 정리된 표지만 그래도 처음 보게 되는 경우는 무슨 얘긴지 알 수 없으니 직접 파일을 열어 위 표와 비교해보면서 확인해보도록 하겠습니다.

![image](images/7.jpg)

위의 이미지를 보면 바이트오더, 0x2A, 0xFF 고정값, 오프셋주소 4바이트 후에 0x0B, 0x00 개의 엔트리 갯수가 "인텔" 방식으로 작성 되었다는 것을 알 수 있습니다. 10진수로 변환하면 11개의 엔트리가 작성되어있다는 것을 의미합니다. 

여기서 아주 중요한 점이 이후 등장하는 <font color="red" style="font-weight:bold;background-color:#eeeeee;">모든 메타 정보의 오프셋 주소는 위의 이미지에서 가르키는 바이트 오더가 시작하는 위치가 오프셋의 기준점</font>이 된다는 사실입니다. "오프셋" 은 보통 어떤 기준점으로부터의 상대적 주소값이라는 의미로 사용됩니다. 예를 들어 위의 APP1 마커에서도 IFD0 오프셋 주소 값이 0x08, 0x00, 0x00, 0x00 으로 8바이트 고정값인데 고정값이 될수 밖에 없는 이유는 2바이트의 바이트 오더값 + 고정값 2바이트 + 오프셋주소 4바이트 = 8바이트 이기 때문에 고정값이 될 수 밖에 없습니다. 위의 이미지에서 처럼 오프셋 기준점에서 8바이트 후를 보면 IFD0 의 엔트리 갯수가 등장하는 위치를 가르킨다는 사실을 알 수 있습니다.

````python
import struct

filename = "note9.jpg"
f = open(filename, "rb")
data = f.read(2)
if data[0:2] != b"\xFF\xD8":
    print("Not JPEG file!!")
    exit(0)

while True:
    H = f.read(2)
    if H == b"\xFF\xE1":
        data = f.read(16)
        length = struct.unpack(">H", data[0:2])[0]
        if data[2:6] != b"Exif":
            print("Not EXIF!")
            exit(0)
        segment_data = f.read(length - 2)
        exif = H + data + segment_data
        break
    elif H == b"":
        break
f.close()

if not exif:
    print("No Exif data")
    exit(0)

HEAD = exif[:10]
TIFF_DATA = exif[10:]
BYTE_ORDER = TIFF_DATA[0:1]
IFD0_ENTRY_COUNT = TIFF_DATA[8:10]
ENDIAN = "<" if BYTE_ORDER == b"I" else ">"
_entry_count = struct.unpack(ENDIAN + "H", IFD0_ENTRY_COUNT)[0]
print("바이트오더: ", BYTE_ORDER)
print("IFD0 엔트리 갯수: ", IFD0_ENTRY_COUNT, _entry_count)
print(TIFF_DATA[:100])
````

기존에 작성한 코드를 보면 우리는 ```exif``` 변수에 모든 정보를 담아두었었습니다. 이 ```exif``` 변수에 저장된 데이터에서 10바이트 만큼은 ```HEAD``` 라는 변수에 저장하고 나머지 모든 데이터는 ```TIFF_DATA``` 라는 변수에 저장했습니다. 


<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
BYTE_ORDER = TIFF_DATA[0:1]
</pre>

여기서 ```TIFF_DATA``` 변수에는 10바이트 이후 바이트 오더가 기록된 부분부터 모든 데이터를 포함하고 있는데 다르게 해석하면 ```TIFF_DATA[0]``` 번째가 <font color="red" style="font-weight:bold;background-color:#eeeeee;">오프셋 주소의 기준점</font>이 된다는 사실 입니다. 그래서 우리는 앞으로 만약 오프셋 주소가 10바이트 만큼이다 라고 하면 ```TIFF_DATA[10:14]``` 이런식으로 쉽게 오프셋 주소를 접근할 수 있게 됩니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
IFD0_ENTRY_COUNT = TIFF_DATA[8:10]
ENDIAN = "<" if BYTE_ORDER == b"I" else ">"
_entry_count = struct.unpack(ENDIAN + "H", IFD0_ENTRY_COUNT)[0]
</pre>

IFD0 의 엔트리 갯수를 구하기 위해선 이제 바이트 오더값을 사용해야 합니다. 위의 코드에서 바이트오더가 "I" 인경우면 ```ENDIAN``` 변수 값은 "<" 가 되고 그렇지 않은 경우는 ">" 값이 됩니다. 이는 ```struct.unpack``` 함수를 사용하기 위함인데 ```struct.unpack(format, buffer)``` 함수는 format 에 따라 buffer 에서 데이터를 언패킹 해주는 함수인데 그냥 쉽게 이해하면 바이트 데이터를 우리가 원하는 형태로 만들어주는 함수라고 보시면 됩니다. 위 코드에서 사이즈 값인 0x0B, 0x00 값은 2바이트 unsigned short 값 이므로 포맷 문자 H 를 사용합니다. [[struct 포맷문자열 참조]](https://docs.python.org/ko/3/library/struct.html#format-strings)

> struct.unpack("<H", b"\x0b\x00")

위와 같이 리틀엔디안 '<', unsigned short 'H' 식으로 작성하면 바이트 오더를 적용하여 0x0B, 0x00 해당하는 값을 우리가 원하는 형태로 돌려주게 됩니다. 주의할 점은 ```unpack``` 함수는 결과를 (11, ) 처럼 튜플 형태로 리턴한다는 사실입니다.

### 엔트리


![image](images/8.jpg)
<center><em style="font-size:12px;">[ 엔트리 정보 ]</em></center>

첫번째로 등장한 엔트리의 태그는 0x12, 0x01 이고 데이터 포맷은 0x03, 0x00, 컴퍼넌트 수는 0x01, 0x00, 0x00, 0x00 이며 데이터 값은 0x01, 0x00, 0x00, 0x00 입니다. 이 데이터는 "인텔" 방식으로 저장되었기 때문에 다르게 표기해보면 태그는 0x0112, 데이터포맷은 0x0003, 컴퍼넌트수는 0x00000001, 데이터값은 0x00000001 입니다.

![image](images/9.jpg)
<center><em style="font-size:12px;">[ 데이터가 아닌 오프셋인경우 ]</em></center>



위의 이미지는 다른 엔트리의 값을 보여주고 있는데 태그는 0x0F, 0x01 로 바이트 정렬을 감안해보면 실제 데이터는 0x010F 이며 태그 뒤에 0x02, 0x00 값은 데이터 포맷이 ascii string 를 의미하며 컴퍼넌트 수는 0x08, 0x00, 0x00, 0x00 이고 실제 데이터는 0xA2, 0x00, 0x00, 0x00 으로 작성된것을 알 수 있습니다. 

> 포맷 바이트 수 * 컴포넌트 갯수 = 실제 데이터 바이트수

인것을 감안했을때 ascii string 은 1바이트와 컴퍼넌트 갯수값 0x00000008 을 10진수로 변경해서 계산하면 1 * 8 = 8 이기 때문에 4바이트를 초과하게 됩니다. 그렇게 되면 <font color="red" style="font-weight:bold">0x000000A2</font> 값은 실제 데이터 값이 아니라 오프셋 값인걸 알 수 있게 됩니다. 

![image](images/10.jpg)
<center><em style="font-size:12px;">[ 데이터가 아닌 오프셋인경우 ]</em></center>

위 이미지에서 처럼 실제 오프셋을 추적해보면 오프셋 시작지점에서 부터 A2 바이트만큼을 드래그하여 선택해보면 오프셋 이후에 실제 데이터가 존재하는것을 확인할 수 있습니다. 오프셋 이후 컴퍼넌트 갯수 0x00000008 만큼의 문자열이 samsung.(점포함 아스키 8글자) 이라고 기록된것을 알 수 있고 해당 엔트리를 정리해보면 카메라 제조사(0x010F) 는 samsung. 이라는 것을 알 수 있게 됩니다.

````python
offset = 10
for i in range(_entry_count):
    pointer = offset + 12 * i
    tag = struct.unpack(ENDIAN + "H", TIFF_DATA[pointer : pointer + 2])[0]
    value_format = struct.unpack(ENDIAN + "H", TIFF_DATA[pointer + 2 : pointer + 4])[0]
    value_num = struct.unpack(ENDIAN + "L", TIFF_DATA[pointer + 4 : pointer + 8])[0]
    value = struct.unpack(ENDIAN + "I", TIFF_DATA[pointer + 8 : pointer + 12])[0]
    print("tag:{:04x} format:{} num:{} value:{}".format(tag, value_format, value_num, value))
````

위의 코드는 IFD0 의 엔트리 갯수만큼 반복문을 돌며 각 엔트리의 태그와 포맷, 컴포넌트 수, 데이터/오프셋 값을 출력하는 코드 입니다. 우리는 모든 데이터를 ```TIFF_DATA``` 변수에 담아두었기 때문에 ```TIFF_DATA``` 변수에 있는 값을 가져오면 되는데 최초 TIFF 헤더 8바이트 + 엔트리 갯수 2바이트, 총 10바이트 후 부터 실제 엔트리의 값이 나타나기 때문에 최초 ```offset=10``` 을 하여 시작 위치를 설정했습니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
pointer = offset + 12 * i
</pre>
각각의 엔트리의 크기는 태그(2바이트) + 포맷(2바이트) + 컴포넌트수(4바이트) + 데이터/오프셋(4바이트) = 12바이트로 규칙적이기 때문에 위의 코드에서 처럼 시작위치 + 12바이트 * 엔트리 수를 반복하면 ```pointer``` 에는 각 엔트리의 시작위치 값이 저장됩니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
tag = struct.unpack(ENDIAN + "H", TIFF_DATA[pointer : pointer + 2])[0]
value_format = struct.unpack(ENDIAN + "H", TIFF_DATA[pointer + 2 : pointer + 4])[0]
value_num = struct.unpack(ENDIAN + "L", TIFF_DATA[pointer + 4 : pointer + 8])[0]
</pre>

태그는 엔트리 시작위치에서 2바이트 만큼이기 때문에 ```TIFF_DATA[pointer : pointer + 2]``` 이렇게 접근할 수 있습니다. 그뒤로 ```TIFF_DATA[pointer + 2 : pointer + 4]``` 는 포맷, ```TIFF_DATA[pointer + 4 : pointer + 8]``` 콤포넌트 갯수, ```TIFF_DATA[pointer + 8 : pointer + 12]``` 데이터 및 오프셋 값을 쉽게 알수 있습니다. 각각의 데이터는 필요에 맞게 ```struct.unpack``` 함수를 통해 변환해서 사용할 수 있습니다. 물론 바이트 오더에 따른 엔디안 값을 항상 염두해야 합니다. 

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
value = struct.unpack(ENDIAN + "I", TIFF_DATA[pointer + 8 : pointer + 12])[0]
</pre>

```value``` 어쨌든 데이터 값이거나 오프셋 주소이기 때문에 일단은 정수형태로 변환하여 저장해놓습니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
print("tag:{:04x} format:{} num:{} value:{}".format(tag, value_format, value_num, value))
</pre>
최종적으로 각 변수에 해당하는 값을 화면에 출력해보면 각각의 엔트리에 해당하는 ```tag:0112 format:3 num:1 value:1``` 이런식의 값을 얻어낼 수 있습니다. 


이제 우리는 ```value``` 의 값이 실제 데이터 값인지 아니면 오프셋 주소 값인지를 알아내야 하는데 그걸 알기 위해선 ```value_format``` 에 해당하는 바이트 값과 ```value_num``` 의 값을 곱해서 실제 데이터가 차지하는 바이트 수를 알아야 합니다. 그 값이 4바이트 이하라면 ```value``` 에는 실제 데이터가 들어가 있을테고 만약 4바이트를 초과하면 ```value``` 변수에는 실제 데이터가 위치하는 오프셋 값이 있을테니 한번 더 오프셋 주소를 찾아서 데이터를 구해야 합니다.

````python
FORMAT_TYPES = (
    (0, ""),
    (1, "B", "unsigned byte"),
    (1, "c", "ascii"),
    (2, "H", "unsigned short"),
    (4, "L", "unsigned long"),
    (8, "Q", "unsigned rational"),
    (1, "b", "signed byte"),
    (1, "", "undefined"),
    (2, "h", "signed short"),
    (4, "l", "signed long"),
    (8, "q", "signed rational"),
    (4, "f", "single float"),
    (8, "d", "double float")
)
````

우리는 [데이터 포맷](#데이터-포맷-(Data-format)) 을 참조하여 데이터 포맷은 1 부터 12까지의 존재하니 0번째 빈 값을 하나 만들어서 위와 같이 튜플 형태로 변수를 작성할 수 있습니다. 이렇게 선언 해놓으면 만약 데이터 포맷 값이 3이라고 한다면 ```FORMAT_TYPES[3]``` 이렇게 접근했을때 우리는 ```(2, "H", "unsigned short")``` 값을 얻을 수 있게 됩니다. 튜플의 첫번째 요소는 해당 포맷의 바이트 수 이고 두번째는 ```struct.unpack``` 함수에서 사용되는 포맷문자열, 그 다음은 우리가 읽을 수 있는 포맷의 이름입니다. 이런 데이터 구조의 형태는 사실 정해진 방법은 없기 때문에 더 좋은 방법을 고민해보고 적용해보는것도 좋은 공부법이기도 합니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
value_size = FORMAT_TYPES[value_format][0] * value_num
</pre>

이제 우리는 ```FORMAT_TYPES[value_format][0]``` 이렇게 ```value_format``` 해당하는 바이트 값을 구할 수 있고 그 값에 ```value_num``` 을 곱해서 실제 데이터 크기를 알아낼 수 있습니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
value_offset = pointer + 8

if value_size > 4:
    value_offset = struct.unpack(ENDIAN + "I", TIFF_DATA[value_offset : value_offset + 4])[0]
</pre>

조금 더 효율적인 방법을 위해 코드를 조금 수정하여 ```value_offset = pointer + 8``` 데이터 / 오프셋 주소 영역의 값을 기본값으로 value_offset 이라는 변수를 선언해서 이제 모든 데이터 접근을 offset 주소로 접근 하는 방식으로 변경합니다. 
위에서 알아낸 ```value_size``` 값이 4 바이트를 초과하는 경우에는 우리는 기본값으로 설정한 ```value_offset``` 변수에 새로운 오프셋 주소값을 구해 저장합니다. ```TIFF_DATA``` 변수의 시작점이 오프셋 기준점이기 때문에 쉽게 ```TIFF_DATA[오프셋:오프셋+데이터크기]``` 를 통해 원하는 오프셋 주소값을 구할 수 있게 됩니다.

````python
FORMAT_TYPES = (
    (0, ""),
    (1, "B", "unsigned byte"),
    (1, "c", "ascii"),
    (2, "H", "unsigned short"),
    (4, "L", "unsigned long"),
    (8, "Q", "unsigned rational"),
    (1, "b", "signed byte"),
    (1, "", "undefined"),
    (2, "h", "signed short"),
    (4, "l", "signed long"),
    (8, "q", "signed rational"),
    (4, "f", "single float"),
    (8, "d", "double float")
)

offset = 10
for i in range(_entry_count):
    pointer = offset + 12 * i
    tag = struct.unpack(ENDIAN + "H", TIFF_DATA[pointer : pointer + 2])[0]
    value_format = struct.unpack(ENDIAN + "H", TIFF_DATA[pointer + 2 : pointer + 4])[0]
    value_num = struct.unpack(ENDIAN + "L", TIFF_DATA[pointer + 4 : pointer + 8])[0]
    value_size = FORMAT_TYPES[value_format][0] * value_num
    value_offset = pointer + 8

    if value_size > 4:
        value_offset = struct.unpack(ENDIAN + "I", TIFF_DATA[value_offset : value_offset + 4])[0]

    value = TIFF_DATA[value_offset : value_offset + value_size]
    print("tag:{:04x} format:{} num:{} value:{} value_size:{}".format(tag, value_format, value_num, value, value_size))
````
