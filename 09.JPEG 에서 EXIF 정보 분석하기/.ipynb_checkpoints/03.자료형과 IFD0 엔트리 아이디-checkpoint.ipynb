### 자료형태에 따른 출력

일단은 현재 value 값이 바이트 형태로 출력되는데 이를 적당한 포맷으로 변경해야 할 필요가 있습니다. 아스키 형태의 문자열인지, short 형태인지 등을 구분하여 변환하는데 여기서 한가지 문제점은 ```value_size > 4``` 보다 큰 경우의 포맷 형태를 보면 2:ascii string, 5:unsigned rational, 10:signed rational, 
11:single float, 12:double float 일때 예외 처리를 해줘야 합니다. ascii string 인 경우에는 그냥 해당 오프셋에서 데이터 사이즈만큼 읽어와서 바로 변환하면 되지만 rational 형태의 경우에 데이터는 실제 8바이트를 사용하지만 앞의 4바이트 뒤의 4바이트 숫자인 분자/분모 형태로 작성되며 float 형태는 single float 일때와 double float 일때 ```struct.unpack``` 함수의 포맷 문자열이 바뀝니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
if value_format == 2:
    ascii string 처리....
elif value_format in (5, 10):
    rational 처리...
elif value_format in (11, 12):
    float 처리...
</pre>

위와 같은 식으로 무언가 작업을 해야 합니다.

### ascii string

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
if value_format == 2:
    value = TIFF_DATA[value_offset : value_offset + value_size].decode()
</pre>

기본적인 ascii string 은 ```decode()``` 함수만으로 변경이 가능하기 때문에 특별한 작업을 할 필요가 없습니다.

### rational

데이터의 포맷이 5 거나 10인 rational 데이터 형태는 대부분 분자/분모 형태로 작성됩니다. 보통 8바이트중 앞의 4바이트는 분자, 뒤의 4바이트는 분모 형태로 작성되기 때문에 우리는 오프셋 주소에서 8바이트를 한번에 읽지 않고 4바이트씩 나눠서 읽어줘야 합니다.
<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
elif value_format in (5, 10):
    value1 = struct.unpack(ENDIAN + "I", TIFF_DATA[value_offset : value_offset + 4])[0]
    value2 = struct.unpack(ENDIAN + "I", TIFF_DATA[value_offset + 4 : value_offset + 8])[0]
    value = "{}/{}".format(value1, value2)
</pre>
위의 코드처럼 앞의 4바이트, 뒤의 4바이트를 읽어서 분자/분모 형태로 출력합니다.

### float
<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
else:
    value = struct.unpack(ENDIAN + FORMAT_TYPES[value_format], TIFF_DATA[value_offset : value_offset + value_size])
</pre>

데이터 포맷이 11 이거나 12 일때 float 형태 인데 여기서 11 은 single float 이고 12인경우에는 double float 이 됩니다. 여기서 11 일때와 12일때 ```struct.unpack()``` 함수의 포맷 문자열이 바뀌기 때문에 이점을 주의해야 합니다. 그러나 우리는 이미 ```FORMAT_TYPES``` 변수에 설정을 해놨기 때문에 포맷이 11 이거나 12인 경우 외에도 다른 형태의 데이터도 위와 같이 간편하게 작성할 수 있습니다.

### 태그 및 기타 값 출력
위에서 작성된 코드를 실행해보면 결과값이 ```tag:0112 format:3 num:1 value:b'\x01\x00' value_size:2``` 이런식으로 출력 되었습니다. 그러면 우리는 IFD0 태그 표를 참고하여 0112 의 태그가 무슨 이름인지 찾아봐야 하고 해당 ```value:x01x00``` 의 값이 무엇을 의미하는지 찾아봐야 합니다. 그래서 이를 좀 더 편하게 하기 위해서는 추가 작업을 해야할 필요가 있습니다. 

````python
IFD0_TAGS = {
    0x010E: ("Image Description", "이미지 설명", {}),
    0x010F: ("Make", "카메라 제조사", {}),
    0x0110: ("Model", "카메라 모델", {}),
    0x0112: ("Orientation", "사진 촬영시 카메라의 방향", {1:"upper-left", 3:"lower-right", 6:"upper-right", 8:"lower-left", 9:"undefined"}),
    0x011A: ("XResolution", "X축 디스플레이/인쇄 해상도", {}),
    0x011B: ("YResolution", "Y축 디스플레이/인쇄 해상도", {}),
    0x0128: ("ResolutionUnit", "위의 디스플레이/인쇄 해상도의 단위", {1:"단위없음", 2:"인치", 3:"센티미터"}),
    0x0131: ("Software", "펌웨어 버전 번호", {}),
    0x0132: ("DateTime", "YYYY:MM:DD HH:MM:SS", {}),
    0x013E: ("WhitePoint", "이미지의 흰색점의 색도 정의.", {}),
    0x013F: ("PrimaryChromaticities", "이미지 원색의 색도 정의", {}),
    0x0211: ("YCbCrCoefficients", "이미지 형식이 YCbCr 인 경우 RGB 형으로 변환하는 상수 표시.", {}),
    0x0213: ("YCbCrPositioning", "이미지 형식이 YCbCr 이고 서브샘플링을 사용하는 경우", {1:"픽셀어레이 중심", 2:"데이텀 포인트 를 의미"}),
    0x0214: ("ReferenceBlackWhite", "흑점/백점의 기준값을 표시", {}),
    0x8298: ("Copyright", "저작권 정보 표시", {}),
    0x8769: ("ExifOffset", "Exif Sub IFD의 오프셋", {}),
}
````

파이썬의 딕셔너리 형태의 변수로 위와 같이 ```IFD0_TAGS``` 이름의 변수를 작성했습니다. 딕셔너리의 키는 태그에 해당하는 16진수 값을 작성하고 키에 해당하는 값은 튜플형태로 해당 태그의 이름, 한글 설명 그리고 또 딕셔너리 형태의 정수 값에 따른 문자열 값으로 설정했습니다. 이런 자료의 형태는 만드는 사람 마음이기 때문에 이런식으로도 작성할 수 있구나 라고 참고하시고 더 좋은 코드를 고민하고 작성해보기시 바랍니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
print("태그:{:04x}, 영문:{}, 한글:{}, 값:{}".format(tag, IFD0_TAGS[tag][0], IFD0_TAGS[tag][1], IFD0_TAGS[tag][2].get(value)))
</pre>

이제 위의 코드처럼 ```IFD0_TAGS[tag]``` 으로 접근하여 [0]번째는 태그의 영문 이름, [1] 번째는 한글설명으로 접근할 수 있으며 해당 태그의 데이터 값이 존재하는 경우 ```IFD0_TAGS[2].get(value)``` 로 접근하여 값이 있는 경우에는 우리가 설정한 값이 출력될테고 없는 경우에는 ```None```의 형태로 출력이 될것 입니다. 

````python
import struct

filename = "note9.jpg"
f = open(filename, "rb")
data = f.read(2)
if data[0:2] != b"\xFF\xD8":
    print("Not JPEG file!!")
    exit(0)

while True:
    H = f.read(2)
    if H == b"\xFF\xE1":
        data = f.read(16)
        length = struct.unpack(">H", data[0:2])[0]
        if data[2:6] != b"Exif":
            print("Not EXIF!")
            exit(0)
        segment_data = f.read(length - 2)
        exif = H + data + segment_data
        break
    elif H == b"":
        break
f.close()

if not exif:
    print("No Exif data")
    exit(0)

HEAD = exif[:10]
TIFF_DATA = exif[10:]
BYTE_ORDER = TIFF_DATA[0:1]
IFD0_ENTRY_COUNT = TIFF_DATA[8:10]
ENDIAN = "<" if BYTE_ORDER == b"I" else ">"
_entry_count = struct.unpack(ENDIAN + "H", IFD0_ENTRY_COUNT)[0]
print("바이트오더: ", BYTE_ORDER)
print("IFD0 엔트리 갯수: ", IFD0_ENTRY_COUNT, _entry_count)
print(TIFF_DATA[:100])

IFD0_TAGS = {
    0x010E: ("Image Description", "이미지 설명", {}),
    0x010F: ("Make", "카메라 제조사", {}),
    0x0110: ("Model", "카메라 모델", {}),
    0x0112: ("Orientation", "사진 촬영시 카메라의 방향", {1:"upper-left", 3:"lower-right", 6:"upper-right", 8:"lower-left", 9:"undefined"}),
    0x011A: ("XResolution", "X축 디스플레이/인쇄 해상도", {}),
    0x011B: ("YResolution", "Y축 디스플레이/인쇄 해상도", {}),
    0x0128: ("ResolutionUnit", "위의 디스플레이/인쇄 해상도의 단위", {1:"단위없음", 2:"인치", 3:"센티미터"}),
    0x0131: ("Software", "펌웨어 버전 번호", {}),
    0x0132: ("DateTime", "YYYY:MM:DD HH:MM:SS", {}),
    0x013E: ("WhitePoint", "이미지의 흰색점의 색도 정의.", {}),
    0x013F: ("PrimaryChromaticities", "이미지 원색의 색도 정의", {}),
    0x0211: ("YCbCrCoefficients", "이미지 형식이 YCbCr 인 경우 RGB 형으로 변환하는 상수 표시.", {}),
    0x0213: ("YCbCrPositioning", "이미지 형식이 YCbCr 이고 서브샘플링을 사용하는 경우", {1:"픽셀어레이 중심", 2:"데이텀 포인트 를 의미"}),
    0x0214: ("ReferenceBlackWhite", "흑점/백점의 기준값을 표시", {}),
    0x8298: ("Copyright", "저작권 정보 표시", {}),
    0x8769: ("ExifOffset", "Exif Sub IFD의 오프셋", {}),
    0x8825: ("GPSInfo", "GPS 정보 IFD 오프셋", {})
}

FORMAT_TYPES = (
    (0, ""),
    (1, "B", "unsigned byte"),
    (1, "c", "ascii"),
    (2, "H", "unsigned short"),
    (4, "L", "unsigned long"),
    (8, "Q", "unsigned rational"),
    (1, "b", "signed byte"),
    (1, "", "undefined"),
    (2, "h", "signed short"),
    (4, "l", "signed long"),
    (8, "q", "signed rational"),
    (4, "f", "single float"),
    (8, "d", "double float")
)

offset = 10
for i in range(_entry_count):
    pointer = offset + 12 * i
    tag = struct.unpack(ENDIAN + "H", TIFF_DATA[pointer : pointer + 2])[0]
    value_format = struct.unpack(ENDIAN + "H", TIFF_DATA[pointer + 2 : pointer + 4])[0]
    value_num = struct.unpack(ENDIAN + "L", TIFF_DATA[pointer + 4 : pointer + 8])[0]
    value_size = FORMAT_TYPES[value_format][0] * value_num
    value_offset = pointer + 8

    if value_size > 4:
        value_offset = struct.unpack(ENDIAN + "I", TIFF_DATA[value_offset : value_offset + 4])[0]

    if value_format == 2:
        value = TIFF_DATA[value_offset : value_offset + value_size].decode()
    elif value_format in (5, 10):
        value1 = struct.unpack(ENDIAN + "I", TIFF_DATA[value_offset : value_offset + 4])[0]
        value2 = struct.unpack(ENDIAN + "I", TIFF_DATA[value_offset + 4 : value_offset + 8])[0]
        value = "{}/{}".format(value1, value2)
    else:
        value = struct.unpack(ENDIAN + FORMAT_TYPES[value_format][1], TIFF_DATA[value_offset : value_offset + value_size])[0]

    value = IFD0_TAGS[tag][2].get(value) if IFD0_TAGS[tag][2].get(value) is not None else value
    print("태그:{:04x}, 영문:{}, 한글:{}, 값:{}".format(tag, IFD0_TAGS[tag][0], IFD0_TAGS[tag][1], value))
````

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
value = IFD0_TAGS[tag][2].get(value) if IFD0_TAGS[tag][2].get(value) is not None else value
</pre>

최종적으로 ```value``` 가 int 인 경우에는 우리가 작성해놓은 ```IFD0_TAGS[tag][2]``` 의 딕셔너리에서 해당 int 형 변수의 값을 구해오는 코드 입니다.

<pre style="background-color:#eeeeee;margin:0px;padding:10px;margin-top:15px;">
IFD0_TAGS = {
    0x0128: ("ResolutionUnit", "위의 디스플레이/인쇄 해상도의 단위", {1:"단위없음", 2:"인치", 3:"센티미터"})
}
</pre>

우리가 작성해놓은 IFD0_TAGS 는 위의 형태로 작성되었기 때문에 IFD0_TAGS[2] 번째가 요소가 ```{1:"단위없음", 2:"인치", 3:"센티미터"}``` 를 가르키게 되고 여기서 ```.get()``` 함수를 통해 1, 2, 3 에 해당하는 키에 따른 값을 구할 수 있게 됩니다. 그래서 만약 딕셔너리의 ```.get()``` 함수를 통해서 값이 존재하면 해당 값을 저장하고 값이 없는 ```None``` 인 경우에는 그냥 ```value``` 변수에 있는 값을 그대로 보관하는 형태 입니다.


### Misc Tags

위의 코드를 실행해보면 맨 마지막 엔트리 ```태그:8825, 영문:GPSInfo, 한글:GPS 정보 IFD 오프셋, 값:833``` 이렇게 출력되는데 이 0x8825 태그는 위에서 작성한 [IFD0 태그 표](#IFD0-태그)에는 존재하지 않는 태그 입니다. 0x8825 태그는 GPS Info 태그로 IFD0 태그가 아닌 Misc Tag 라고 기타 태그에 포함되는 태그입니다. 실제 제대로 된 EXIF 분석 프로그램을 작성하려면 이런 모든 태그에 대해 대처할 수 있을만큼 완성도 있게 프로그램을 만들어야 합니다. 어쨌든 이런 태그에 대해서 모두 다룰 수는 없으니 [[EXIF TAGS LIST]](https://exiftool.org/TagNames/EXIF.html) 링크를 참고해보시기 바랍니다.
