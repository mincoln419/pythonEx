# JPEG

우리가 디지탈카메라 등에서 흔하게 사용하는 확장자가 **.jpg, .jpeg, .jpe** 파일인 이미지 포맷은 정지 화상을 위해 만들어진 손실 압축 방법 표준의 형식입니다. JPEG은 이미지를 어떻게 압축하는가에 대한 방식을 말하고 이 JPEG 로 압축된 데이터를 JFIF 형태로 저장하면 우리가 말하는 .jpg, .jpeg 파일이 되는거라고 볼 수 있습니다.

위 표는 [[위키피아의 JPEG 구조표]](https://en.wikipedia.org/wiki/JPEG#Syntax_and_structure) 내용을 재작성한 내용입니다. 우리는 평상시 jpg 파일 하면 그냥 이미지를 떠올렸습니다만 실제 jpg 파일은 위의 표에서처럼 우리가 평상시 듣도 보도 못한 용어들의 이름으로 복잡하게 구성되어 있으며 jpg 뷰어가 이런 데이터를 분석하고 디코딩하여 우리에게 이미지를 보여주게 됩니다. 이 강좌에선 우리는 JPEG 파일의 전체 구조를 분석하고 이해하려는 목적은 아니고 위의 표에서 붉은 색으로 표시된 EXIF 정보에 대해서만 다룰 예정입니다. 


### EXIF (<font color="red">EX</font>changeable <font color="red">I</font>mage <font color="red">F</font>ile format)

디지털 카메라 등에서 사용되는 이미지 파일 메타데이터 포맷으로 가장 중요한 특징은 카메라의 기종, 촬영시간, GPS 정보, 초점거리, 플래시 발광여부 등등등 수많은 메타데이터를 기록할 수 있다는 점 입니다. 이런 EXIF 정보는 꼭 JPEG 파일만 사용되는것은 아니고 TIFF 6.0, RIFF, WAV 파일 그리고 PNG (v1.5 버전이상) 등 여러 포맷의 일부로 저장되는 메타데이터 포맷입니다.

![image](images/exif.jpg)

이미지에 EXIF 정보가 존재한다면 이미지 뷰어 프로그램을 사용하여도 EXIF 정보를 볼 수 있지만 위 처럼 윈도우 운영체제의 탐색기에서 이미지 파일을 우클릭하여 속성만 살펴보아도 간단하게 EXIF 정보를 살펴볼 수 있습니다. EXIF 정보는 사진에 대한 추가 정보를 많이 담고 있기 때문에 이미지 외에도 여러가지 데이터를 동시 기록한다는 장점도 있지만 GPS 정보같이 민감한 데이터는 개인 정보 유출 혹은 사생활 침해도 될 수 있기 때문에 좀 더 각별하게 신경을 써야 하는 부분도 있습니다. 참고로 대부분의 스마트폰 카메라의 GPS 옵션은 디폴트 값이 False 입니다.

![image](images/jpeg_structure.jpg)

EXIF 포맷은 일본전자정보 기술산업협회(JEITA)에 의해 개발되었는데 위 이미지는 [[JEITA Exif 2.2 문서]](http://exif.org/Exif2-2.PDF)에서 어플리케이션 마커(APP1) 일부의 내용을 재작성한 표 입니다. 강좌 처음의 표에서 APP<em style="font-weight:bold;">n</em> 으로 기술된 부분입니다.

어떻게 보면 정말 잘 설명해놓은 표 이미지지만 태어나서 처음 보는 입장에선 무슨 얘긴지 도통 알수가 없습니다. 문제는 우리가 필요로 하는 EXIF 데이터는 위의 이미지에서 처럼 일반적으로 APP1 안에 기록되는건 알겠는데 APP1도 단순하게 생긴거 같지가 않습니다. 어떻게 저 APP1 에 접근하고 또 정보를 얻어와야 하는지 아무것도 모르니 일단 16진수 뷰어/편집기 를 통해 JPEG 파일부터 열여보는걸로 시작해보도록 하겠습니다. 16진수 뷰어는 이전 MP3 강좌에서도 사용했지만 각자의 운영체제에 맞는 프로그램을 찾으셔야 하고 윈도우 운영체제 같은경우 [[Hxd]](https://mh-nexus.de/en/downloads.php?product=HxD20) 를 추천합니다.


![image](images/1.jpg)

위의 이미지는 일반적인 스마트폰으로 촬영한 jpg 파일을 16진수 뷰어로 본 모습 입니다. 위의 [JPEG 구조표](#JPEG-구조)에서 처럼 JPEG 파일의 시작 (SOI)은 항상 0xFF, 0xD8 로 시작하는것을 볼 수 있습니다. 그리고 얼핏 보면 우리가 알아볼 수 있는 글자도 보이는것도 알 수 있습니다.

![image](images/2.jpg)

JPEG 이미지의 시작임을 알리는 0xFF, 0xD8 바로 뒤에는 0xFF, 0xE1 값이 바로 자리 하고 있습니다. 이 부분이 위의 [JPEG 구조표](#JPEG-구조)에서 빨간색 박스로 강조해놓은 App<em style="font-weight:bold;">n</em> 영역의 마커 표시인데 0xFF, 0xE1 은 APP1 의 마커이고 우리가 필요로하는 EXIF 정보가 이곳에 담겨 있습니다. 여기서 간과해서 안되는 부분은 항상 위의 상황처럼 0xFF, 0xD8 바로 뒤에 APP1 의 값이 존재한다는것은 아닙니다. 만약 이미지에 JFIF 정보가 있다면 JFIF 정보는 APP0 을 사용하기 때문에 APP1 보다 먼저 등장할 수도 있습니다.

![image](images/4.jpg)

APP1 마커 뒤에는 사실 APP1 의 데이터 크기가 표기됩니다만 위와 같이 그냥 최대값 0xFF, 0xFF 로 기재되는 경우가 있습니다. 그렇기 때문에 이 값을 실제 사용하기에는 무리가 있을 수 있습니다.

![image](images/3.jpg)

EXIF 헤더는 위의 이미지에서 처럼 "Exif" 문자열을 나타내는 16진수 값인 0x45(E), 0x78(x), 0x69(i), 0x66(f) 그리고 Null 문자 2바이트 0x00, 0x00 로 고정값이라보 보면 됩니다. 

![image](images/5.jpg)

Exif 헤더 바로 뒤에는 8바이트 TIFF 헤더가 위치합니다. 앞에서부터 살펴보면 앞의 2바이트는 데이터의 바이트 정렬을 정의합니다. 위의 이미지에서는 0x49, 0x49 는 아스키문자로 II 로 표기되는데 "Intel" 유형의 바이트 정렬을 의미하고 0x4D, 0x4D 는 아스키문자로 MM 으로 표기되는데 이는 "Motorola" 유형의 바이트 정렬을 의미합니다. 인텔방식은 리틀엔디안, 모토로라 방식은 빅엔디안을 사용 하는데 쉽게 리틀엔디안은 작은수를 오른쪽부터, 빅엔디안은 큰수를 오른쪽부터 기입하는 방식이라고 보시면 됩니다. 예를 들어 0x123456 이라는 값을 표기할때 리틀엔디안(II) 은 0x56, 0x34, 0x12 로 기입하고 빅엔디안은 0x12, 0x34, 0x56 으로 기입하게 됩니다. 이미지 파일내의 모든 EXIF 데이터를 분석할때 이 바이트 정렬 방식을 기준으로 하기 때문에 상당히 중요한 부분입니다.

![image](images/6.jpg)

위의 이미지는 빅엔디안 방식 "MM" 형식으로 모토로라 방식의 바이트 정렬의 샘플 이미지 입니다. 바이트정렬 다음 2바이트는 0x2A, 0x00 고정값으로 만약 위의 이미지처럼 모토롤라 방식이였다면 0x00, 0x2A 로 작성되게 됩니다. 그리고 마지막 4바이트는 첫번째 IFD 에 대한 오프셋인데 현재 TIFF 헤더값이 8바이트로 작성되어있으니 첫번째 IFD 오프셋 값은 8바이트 고정값이 됩니다. 오프셋은 현재 위치에서부터 상대적 주소값 입니다.

실제 데이터 총 바이트 길이는 위 표의 "바이트/콤포넌트" * 컴포넌트 갯수 를 계산하여 얻을 수 있습니다. 이렇게 계산하여 얻어진 바이트 길이가 4바이트 미만이면 실제 데이터는 엔트리의 데이터/오프셋 영역에 작성되지만 4바이트를 넘어가게 되면 데이터/오프셋 영역에는 실제 데이터가 위치하는 오프셋 주소값만 기록되게 됩니다. 위의 표들이 간단하게 정리된 표지만 그래도 처음 보게 되는 경우는 무슨 얘긴지 알 수 없으니 직접 파일을 열어 위 표와 비교해보면서 확인해보도록 하겠습니다.

![image](images/7.jpg)

위의 이미지를 보면 이 이미지 파일은 0x0B, 0x00 개의 엔트리 갯수가 "인텔" 방식으로 작성 되었다는 것을 알 수 있습니다. 10진수로 변환하면 11개의 엔트리가 작성되어있다는 것을 의미합니다.

````python
import struct

filename = "note9.jpg"
f = open(filename, "rb")
data = f.read(2)
if data[0:2] != b"\xFF\xD8":
    print("Not JPEG file!!")
    exit(0)

while True:
    H = f.read(2)
    if H == b"\xFF\xE1":
        data = f.read(16)
        length = struct.unpack(">H", data[0:2])[0]
        if data[2:6] != b"Exif":
            print("Not EXIF!")
            exit(0)
        segment_data = f.read(length - 2)
        exif = H + data + segment_data
        break
    elif H == b"":
        break
f.close()

if not exif:
    print("No Exif data")
    exit(0)

HEAD = exif[:10]
TIFF_DATA = exif[10:]
BYTE_ORDER = TIFF_DATA[0:1]
IFD0_ENTRY_COUNT = TIFF_DATA[8:10]

print("바이트오더: ", BYTE_ORDER)
print("IFD0 엔트리 갯수: ", IFD0_ENTRY_COUNT)
print(TIFF_DATA[:100])
````

기존에 작성한 코드를 보면 우리는 ```exif``` 변수에 모든 정보를 담아두었었습니다. 이 ```exif``` 변수에 저장된 데이터에서 10바이트 만큼은 ```HEAD``` 라는 변수에 저장하고 나머지 모든 데이터는 ```TIFF_DATA``` 라는 변수에 저장했습니다. 여기서 ```TIFF_DATA``` 변수에는 10바이트 이후 바이트 오더가 기록된 부분부터 모든 데이터를 포함하고 있는데 다르게 해석하면 ```TIFF_DATA[0]``` 번째가 <font color="red" style="font-weight:bold;background-color:#eeeeee;">오프셋 주소의 기준점</font>이 된다는 사실 입니다. 그래서 우리는 앞으로 만약 오프셋 주소가 10바이트 만큼이다 라고 하면 ```TIFF_DATA[10:14]``` 이런식으로 쉽게 오프셋 주소를 접근할 수 있게 됩니다.

위의 표를 참조하여 태그 0x0112 는 사진이 촬영될때 카메라의 방향을 의미하는 Orientation 을 의미한다는 것을 알 수 있습니다. 예제 이미지에서 실제 데이터 0x00000001 의 데이터 포맷 0x0003 은 unsigned short 의 자료형을 의미하기 때문에 값은 그냥 1이며 이 값은 사진이 촬영될때 카메라는 좌상단 방향이였다는걸 알 수 있습니다.


![image](images/8.jpg)
<center><em style="font-size:12px;">[ 데이터가 아닌 오프셋인경우 ]</em></center>

위의 이미지는 다른 엔트리의 값을 보여주고 있는데 태그는 0x0F, 0x01 로 바이트 정렬을 감안해보면 실제 데이터는 0x010F 이며 위의 IFD0 태그를 살펴보면 0x010F는 Make 로 카메라의 제조사 정보라는 것을 알 수 있습니다. 태그 뒤에 0x02, 0x00 값은 데이터 포맷이 ascii string 를 의미하며 컴퍼넌트 수는 0x08, 0x00, 0x00, 0x00 이고 실제 데이터는 0xA2, 0x00, 0x00, 0x00 으로 작성된것을 알 수 있습니다. 

> 포맷 바이트 수 * 컴포넌트 갯수 = 실제 데이터 바이트수

인것을 감안했을때 ascii string 은 1바이트와 컴퍼넌트 갯수값 0x00000008 을 10진수로 변경해서 계산하면 1 * 8 = 8 이기 때문에 4바이트를 초과하게 됩니다. 그렇게 되면 <font color="red" style="font-weight:bold">0x000000A2</font> 값은 실제 데이터 값이 아니라 오프셋 값인걸 알 수 있게 됩니다. 

![image](images/10.jpg)
<center><em style="font-size:12px;">[ 데이터가 아닌 오프셋인경우 ]</em></center>

위 이미지에서 처럼 실제 오프셋을 추적해보면 오프셋 시작지점에서 부터 A2 바이트만큼을 드래그하여 선택해보면 오프셋 이후에 실제 데이터가 존재하는것을 확인할 수 있습니다. 오프셋 이후 컴퍼넌트 갯수 0x00000008 만큼의 문자열이 samsung.(점포함 아스키 8글자) 이라고 기록된것을 알 수 있고 해당 엔트리를 정리해보면 카메라 제조사(0x010F) 는 samsung. 이라는 것을 알 수 있게 됩니다.

![image](images/11.jpg)

이런식으로 IFD0 의 엔트리를 순환하여 모든 값을 추출할 수 있고 최종적으로 IFD0 표에서 0x69, 0x87 의 태그는 Exif IFD 의 오프셋을 가르킨다고 했는데 이 값이 만약 0x00, 0x00, 0x00, 0x00 이면 해당 IFD 가 존재 하지 않는다고 판단하면 됩니다. 어쨌든 위의 이미지를 보면 Exif IFD 의 오프셋이 0xD5, 0x00, 0x00, 0x00 으로 작성된것을 확인할 수 있습니다. 그렇다면 이 이미지는 IFD0 외에 Exif IFD 가 또 존재한다는 것을 알 수 있으며 지금까지와 같은 방식으로 Exif IFD 의 엔트리를 분석할 수 있습니다.

![image](images/12.jpg)

위의 이미지에서 처럼 Exif IFD 오프셋을 가르키는 0x69, 0x87 태그의 데이터로 작성된 오프셋 값 0xD5, 0x00, 0x00, 0x00 만큼을 오프셋 시작지점부터 드래그 해보면 Exif IFD 가 시작되는 지점을 알 수 있습니다. 이제 이 부분부터는 IFD0 이 아닌 Exif IFD 영역으로 넘어가게 됩니다.

![image](images/13.jpg)

Exif IFD 역시 최초 2바이트는 Exif IFD 의 엔트리 갯수가 작성되어 있습니다. 위의 이미지에서를 보면 현재 이미지의 Exif IFD 에는 0x0024(십진수 36)개의 엔트리가 존재한다는 것을 알 수 있습니다.

![image](images/14.jpg)

Exif IFD 의 엔트리는 IFD0 의 엔트리와 동일한 구조로 작성되어있습니다. 위의 이미지를 보면 태그는 0x9A, 0x82 이고 데이터포맷은 0x05, 0x00, 컴퍼넌트 수는 0x01, 0x00, 0x00, 0x00, 데이터는 0x8B, 0x02, 0x00, 0x00 으로 되어있습니다. 이를 바이트 정렬을 감안하여 풀이해보면 태그는 0x829A, 포맷은 0x0005 이므로 unsigned rational 형태이고 컴퍼넌트 갯수는 1개 입니다. 그런데 이 unsigned rational 데이터 포맷은 컴포넌트당 8 바이트를 사용하기 때문에 이 엔트리의 데이터 필드에 작성된 0x0000028B 값은 실제 값이 아닌 오프셋 주소 값이라고 봐야 합니다.

![image](images/15.jpg)

위 엔트리에서 0x028B 이 오프셋 주소라 했으니 아까 IFD0 에서 오프셋 시작 주소부터 0x028B 만큼을 이동해보 unsigned rational 포맷 8바이트 만큼을 살펴보면 0x01, 0x00, 0x00, 0x00, 0x04, 0x00, 0x00, 0x00 이라는 데이터가 작성되어있는것을 확인할 수 있습니다. 결론을 정리해보면 Exif IFD 의 첫번째 엔트리는 0x829A 태그에 0x01, 0x00, 0x00, 0x00, 0x04, 0x00, 0x00, 0x00 라는 데이터가 작성되었다고 볼 수 있습니다.

위의 GPS 태그 표를 보면 방금전의 첫번째 엔트리는 0x0000 의 태그는 GPS 버전을 의미하고 데이터는 0x0202 는 GPS 버전 기본값인 2.2 버전인것을 알 수 있습니다. 

![image](images/18.jpg)

다음 엔트리를 살펴보면 0x01, 0x00 태그로 북위 또는 남위를 의미하고 값은 "N" 인것을 확인할 수 있습니다. 그 뒤에 나오는 0x02, 0x00 엔트리는 위도 엔트리인데 unsinged rational 포맷의 컴퍼넌트가 3개라고 되어있습니다. unsigned rational 은 컴퍼넌트 당 8바이트를 차지하기 때문에 8 * 3 = 24 바이트 만큼을 오프셋 주소에서 읽어야 합니다.

![image](images/19.jpg)

위도 엔트리의 데이터를 구하기 위해 실제 오프셋 위치(**0x03BF**)에서 24바이트를 살펴보면 위와 같이 기록되어있습니다. GPS 데이터 위도 경도의 3개의 컴포넌트는 각각 도, 분, 초의 형식으로 작성되어있으며 일반적으로 도/1, 분/1, 초/1 형식으로 작성됩니다. 따라서 첫번째 8바이트를 보면 0x24, 0x00, 0x00, 0x00, 0x01, 0x00, 0x00, 0x00 으로 작성되어있는데 이 데이터는 4바이트/4바이트 형식으로 이해하고. 0x00000024 / 0x00000001 로 볼 수 있습니다. 이를 십진수로 변환하면 36/1 로 표기될 수 있습니다. 두번째, 세번째 모두 이런식으로 변환하면 36/1, 53/1, 60/1 로 작성할 수 있고 해당 이미지의 GPS 위도 좌표는 36°53'60.00" 로 이해할 수 있게 됩니다. 이 표기법은 도,분,초로 표기한 방식인데 구글맵 같은데서 위도를 검색할때는 10진수 표기법으로 작성되야 합니다. 이 도,분,초 형식을 10진수 표기법으로 바꾸려면

> 도 + ((분 / 60) + (초 / 3600)) = 10진법 좌표

위의 식을 이용하여 우리가 구한 도,분, 초의 값에 대입해보면 36 + ((53 / 60) + (60 / 3600)) = 36.9 이와 같은 결과를 얻을 수 있습니다. 이런방식으로 GPS IFD 안의 엔트리 정보도 모두 구할 수 있습니다.

JPEG 안에는 현재 노트에 작성된 IFD0, Exif IFD 태그 외에도 썸네일 이미지를 저장하는 IFD1 외에도 카메라 제조사마다 사용하는 독립적인 IFD 가 따로 또 존재합니다. 그렇기 때문에 완벽한 Exif 분석 프로그램을 만들기 위해선 이런 태그를 알고 작성하는데 많은 시간을 투자해야 합니다. 좀 더 자세한 태그 정보는 [[EXIF 2.2 PDF 문서]](http://exif.org/Exif2-2.PDF) 혹은 [[Description of Exif file format]](https://www.media.mit.edu/pia/Research/deepview/exif.html) 링크를 참고하시기 바랍니다.

### IFD1 (Thumbnail Image)

![image](images/20.jpg)

위의 표는 IFD1 의 태그 정보 입니다. IFD1 역시 IFD0 과 동일한 방식으로 엔트리 갯수를 구하고 각각의 엔트리에 대한 정보를 위의 표를 참조하여 구할 수 있습니다.

![image](images/22.jpg)

예를 들어 위의 이미지에서의 엔트리는 0x0201 (JpegIFOffset) 으로 썸네일이 JPEG 이미지인 경우 실제 썸네일 JPEG 데이터가 저장된 곳의 오프셋 이 담긴 엔트리의 정보 입니다.

![image](images/23.jpg)

오프셋 주소만큼 이동해서 살펴보면 실제 썸네일 이미지의 SOI 값인 0xFF, 0xFD 를 확인할 수 있으니 실제 이미지 데이터가 여기서부터 시작한다는 것을 알 수 있습니다.

![image](images/24.jpg)

방금 0x0201 의 태그 바로 다음 엔트리를 보면 태그가 0x0202 (JpegIFByteCount)로 실제 썸네일 이미지 데이터의 바이트수가 기록되어 있습니다. 위의 이미지를 보면 실제 데이터는 0x297B 라고 되어있습니다. 

![image](images/25.jpg)

썸네일 이미지의 실제 데이터가 시작하는 SOI 부분부터 위의 0x0202 에 작성된 실제 썸네일 이미지의 데이터 바이트 수인 0x297B 만큼을 드래그 해보면 실제 썸네일 데이터의 EOI 를 확인 할 수 있습니다.

### Misc Tags

위의 코드의 맨 마지막 출력값을 보면 ```tag:8825 format:4 num:1 value:833 value_size:4``` 이런 값이 찍히는데 문제는 0x8825 태그라는게 [IFD0 태그 표](#IFD0-태그)에는 존재하지 않습니다. 0x8825 태그는 GPS Info 태그로 IFD0 태그가 아닌 Misc Tag 라고 기타 태그에 포함되는 태그입니다. 사실 이 외에도 수많은 태그들이 존재합니다만 여기서 모두 다루지는 않고 링크로 대체하겠습니다.
