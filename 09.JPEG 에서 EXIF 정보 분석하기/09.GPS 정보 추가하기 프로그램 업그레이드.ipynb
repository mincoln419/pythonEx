### GPS 기록 추가하기 프로그램 업그레이드

기존에 작성한 함수형 프로그램의 사용 편리성을 위해 업그레이드

```python
import os
import threading
import pyperclip        # pip install pyperclip
import webbrowser
import time
import re
import sys
from PIL import Image
import piexif
import winreg
import win32com.shell.shell as shell

g_lat = 0
g_lon = 0

class WriteGPS():
    def __init__(self, filename, lat, lon):
        self.filename = filename
        self.latitude = self._10to60(lat) if lat > 0 else None
        self.longitude = self._10to60(lon) if lon > 0 else None
    def _10to60(self, num):
        # 37.59093347188114, 126.95747355774975
        do = int(num)
        bun = int((num - do) * 60)
        cho = int((((num - do) * 60) - bun) * 60)
        return ((do, 1), (bun, 1), (cho, 1))
    def write_gps(self):
        if self.latitude is not None and self.longitude is not None:
            image = Image.open(self.filename)
            exif_dict = piexif.load(image.info.get("exif")) if image.info.get("exif") is not None else {}
            gps_ifd = {
                piexif.GPSIFD.GPSVersionID: (2, 0, 0, 0),
                piexif.GPSIFD.GPSLatitudeRef: "N",
                piexif.GPSIFD.GPSLatitude: self.latitude,
                piexif.GPSIFD.GPSLongitudeRef: "E",
                piexif.GPSIFD.GPSLongitude: self.longitude,
                piexif.GPSIFD.GPSDateStamp: "1999:99:99 99:99:99",
            }
            # if exif_dict.get("GPS") is None:
            #     exif_dict["GPS"] = gps_ifd
            exif_dict["GPS"] = gps_ifd
            exif_bytes = piexif.dump(exif_dict)
            image.save(self.filename, "jpeg", exif=exif_bytes)

class MonitorClipboard(threading.Thread):
    def __init__(self):
        super().__init__()
        self.daemon = True
        self.last_data = None
        self.do_run = True
    def __isNumber(self, s):
        try:
            float(s)
            return True
        except ValueError:
            return False
    def run(self):
        global g_lat, g_lon
        while self.do_run:
            try:
                data = pyperclip.paste()
                if data != self.last_data and data.find(",") > 0:
                    print("클립보드 변경 감지 >>>> " + data)
                    numbers = data.split(",")
                    if self.__isNumber(numbers[0]) and self.__isNumber(numbers[1]):
                        g_lat = float(numbers[0])
                        g_lon = float(numbers[1])
                        self.do_run = False
                        break
                    # match = re.match(r'(^[+-]?\d*(\.\d*))', data)
                    # if match is not None:
                    #     g_lat = float(match.group(1))
                    #     g_lon = float(match.group(2))
                    #     self.do_run = False
                    #     break
                    self.last_data = data
            except Exception as e:
                print(e)
            time.sleep(0.3)

if __name__ == "__main__":
    if len(sys.argv) >= 2:
        filename = sys.argv[1]
        if os.path.exists(filename):
            mon = MonitorClipboard()
            mon.start()
            webbrowser.open("https://www.google.com/maps/")
            mon.join()
            print("Lat:{} Lon:{}".format(g_lat, g_lon))
            _w = WriteGPS(filename, g_lat, g_lon)
            _w.write_gps()
        elif filename == "--add":
            if sys.argv[-1] != "asadmin":
                script = os.path.abspath(sys.argv[0])
                params = " ".join([script] + sys.argv[1:] + ["asadmin"])
                
                # ShellExecuteEx 함수는 프로그램을 실행시킬수 있는 윈도우 API 함수
                # lpVerb : open, edit, explore, find, properties, runas 등의 옵션 여기서 관리자 권한 실행을 위해선 runas 설정
                shell.ShellExecuteEx(lpVerb="runas", lpFile=sys.executable, lpParameters=params)
                sys.exit(0)
            key = winreg.CreateKey(winreg.HKEY_CLASSES_ROOT, "SystemFileAssociations\\.jpg\\Shell\\WriteGPS")
            winreg.SetValue(key, "", winreg.REG_SZ, "GPS 기록하기(&W)")
            winreg.CloseKey(key)
            key = winreg.CreateKey(winreg.HKEY_CLASSES_ROOT, "SystemFileAssociations\\.jpg\\Shell\\WriteGPS\\command")
            winreg.SetValue(key, "", winreg.REG_SZ, '"c:\\Python39\\python.exe" "{}" "%1"'.format(os.path.abspath(__file__)))
            winreg.CloseKey(key)
```
